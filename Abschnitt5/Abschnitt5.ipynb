<h2 style="display:none">Lektion 1 - Lifecycle</h2>
<p>In Android besitzt jede Aktivität und jedes Fragment einen eigenen Lebenszyklus. Dieser beschreibt den aktuellen Status des Elements und kann als Zustandsdiagramm dargestellt werden. In dieser Lektion liegt der Fokus auf der Aktivität. Bei einer Änderung des Zustandes werden bestimmte Methoden aufgerufen, die überschrieben werden können. Eine bereits kennengelernte ist <i>onCreate()</i>. Diese wird aufgerufen, wenn die Aktivität gestartet neu wird. Abgesehen von dieser gibt es aber weitere Zustände als <i>gestartet</i> und <i>zerstört</i>. Diese sollen Sie mit Hilfe von <i>Timber</i> selbst herausfinden, bevor sie in einem Schema vorgestellt werden. Überschreiben Sie dafür folgende Methoden in der Klasse <i>MainAcitivity</i> und fügen Sie ihnen aussagekräftige Logeinträge hinzu:</p>
<ul>
    <li>onStart()</li>
    <li>onResume()</li>
    <li>onRestart()</li>
    <li>onPause()</li>
    <li>onStop()</li>
    <li>onDestroy()</li>
</ul>

<details>

<p>Wie Sie gerade gesehen haben, existieren viele verschiedene Lebenszustände, die alle eine bestimmten Bereich abdecken und in dem Schaubild rechts vereinfacht dargestellt sind. Zusätzlich haben Sie feststellen können, dass immer Paare von Methoden einem Zustand zugeordnet werden können. Ein Beispiel dafür ist <i>onStart()</i> und <i>onStop()</i>. Alle Aktionen, die in <i>onStart()</i> ausgelöst werden, sollten im Gegenstück <i>onPause()</i> auch wieder gestoppt werden.<br />
Bis jetzt wurden alle Aktionen in <i>onCreate()</i> gestartet. Dies ist auch in der Regel so, jedoch ist es beispielsweise bei ressourcenfressenden Simulationen sinnvoll diese in <i>onResume()</i> starten zu lassen, da diese sonst auch berechnet werden könnten, wenn die App auf dem <i>Back Stack</i> liegt.<br />
Besonders hervorgehoben soll noch die linke Hälfte der Abbildung werden. Wenn zu viele Apps im Hintergrund (<i>Back Stack</i>) geöffnet sind, kann die Rechenleistung für die im Vordergrund befindliche App eingeschränkt werden. Dann schließt das Betriebssystem automatisch Apps im Hintergrund. Diese werden jedoch weiterhin im <i>Back Stack</i> angezeigt. Wird die App aus diesem wieder geöffnet, wird wieder <i>onCreate()</i> aufgerufen.</p>

<h2 style="display:none">Lektion 2 - ViewModel</h2>

<p>Zu einer guten Architektur gehört die Trennung von <i>View</i> und <i>Data</i>. Aktuell werden alle Daten im Fragment oder in der Aktivität gespeichert und angepasst. Deswegen wird eine neue Klasse eingeführt, die für die Verarbeitung und Kalkulation der Daten zuständig ist. Ein weiterer Grund für die Trennung ist das Verhalten bei einem abrupten Schließen der App durch das Betriebssystem. Da davor nicht <i>onDestroy()</i> aufgerufen und die App geregelt beendet wird, können Daten verloren gehen. Eine Möglichkeit dies in dem Fragment oder der Aktivität direkt zu lösen, ist die Verwendung einer weiteren <i>Lifecycle</i>-Methode mit dem Namen <i>onSavedInstanceState</i>, der die vorher gespeicherten Daten übergeben und in dieser wiederhergestellt werden können. Eine elegantere Lösung ist die Verwendung eines <i>ViewModels</i>, dass unabhängig von der Aktivität oder dem Fragment besteht. In dieses werden alle Daten und Methoden ausgelagert, die der Ansicht nicht direkt zugeordnet werden. Das Fragment oder die Aktivität erzeugt dabei bei Start das Model und kann somit auf dieses zugreifen. Der Weg vom Model zur <i>View</i> ist jedoch nicht direkt navigierbar. Es wird nun die Architektur nach folgendem Plan implementiert:</p>

<img src="images/Architektur1.png" />

<p>Die Klasse, die aktuell die Daten beinhaltet, ist das <i>MainFragment</i>. Deswegen wird für diese ein <i>ViewModel</i> erstellt. Verschieben Sie die Klasse zuerst in das neue zu erstellende Paket <code>main</code>, welches sich in <i>fragments</i> befinden soll. Erstellen Sie in dem Paket eine neue Klasse <code>MainViewModel</code>, die von <code>ViewModel</code> erbt. Als Vorgriff auf Lektion 3 soll ein Objekt der Klasse mit dem Entwurfsmuster <a target="_blank" href="https://de.wikipedia.org/wiki/Fabrikmethode"><i>Factory</i></a> erzeugt werden. Fügen Sie dem Paket dafür eine weitere Klasse mit dem Namen <code>MainViewModelFactory</code> hinzu.<br />
Bevor die Fabrik implementiert werden kann, muss der Konstruktor des <i>ViewModels</i> definiert werden. Um später auf die Datenbanken zugreifen zu können, muss dem Model der Kontext der App übergeben werden. Fügen Sie dem Konstruktor dafür einen Parameter <code>application</code> der Klasse <code>Application</code> hinzu.<br />
Die <i>Factory</i> verwendet zur Erstellung eines <i>ViewModel</i>-Objekts die bereits in Android vorhandene Schnittstelle <code>ViewModelProvider.Factory</code>. Nur so wird das Model dem Lebenszyklus des Fragments hinzuzufügt. Dem Konstruktor der Fabrik muss ebenfalls eine <i>Application</i> übergeben werden. Daraus ergibt sich folgender Code:</p>

<blockquote><pre><code>class MainViewModelFactory(
    private val application: Application
) : ViewModelProvider.Factory {
    @Suppress("unchecked_cast")
    override fun &#60;T : ViewModel?> create(modelClass: Class&#60;T>): T {
        if (modelClass.isAssignableFrom(MainViewModel::class.java)) {
            return MainViewModel(application) as T
        }
        throw IllegalArgumentException("Unknown ViewModel class")
    }
}</code><pre></blockquote>
    
<p>Mit dem Überschreiben der Methode <i>create()</i> kann das <i>ViewModel</i> erzeugt werden. Die Klasse des Models ist in der <i>if</i>-Verzweigung angegeben. Zurückgegeben wird ein Objekt des Models, dem alle Parameter übergeben wurden.<br />
    Deklarieren Sie in der Klasse <i>MainFragment</i> eine private, veränderbare Variable <code>viewModel</code> der Klasse <code>MainViewModel</code>. Durch diese Referenz ist es der Ansicht möglich Daten zu übertragen oder Aktionen des Models anzustoßen. Die Initialisierung erfolgt dabei in <i>onCreate()</i>:</p>

<blockquote><pre><code>val application = requireNotNull(this.activity).application
val viewModelFactory = MainViewModelFactory(application)
viewModel = ViewModelProvider(this, viewModelFactory).get(MainViewModel::class.java)</code><pre></blockquote>

<p>Wenn Sie nun die App starten, sollten Sie keine Veränderung feststellen, da nichts weiter verändert wurde. Bevor weitere Elemente der Architektur hinzugefügt werden, können Sie in diesem <a target="_blank" href="https://github.com/Limatationz/kotlin-mooc-App-DrinkDrinkDrink/compare/grundgeruest...schritt01-viewModel">git diff</a> die Änderungen betrachten.</p>

<h3>LiveData</h3>

<img src="images/Architektur2.png" style="float:right; margin-left: 10px" />

<p>Die Verbindung vom Model zur Ansicht erfolgt über einen Umweg. Anstelle einer Referenzierung, wie von der Ansicht zum Model, wird das Entwurfsmuster <a target="_blank" href="https://de.wikipedia.org/wiki/Beobachter_(Entwurfsmuster)"><i>Observer</i></a> verwendet. Das Fragment beobachtet die Daten des <i>ViewModels</i> und wird bei Veränderung benachrichtigt. Diese Daten müssen aber das Entwurfsmuster unterstützen. Deswegen wurde die Klasse <i>LiveData</i>, beziehungsweise <i>MutableLiveData</i> eingeführt. Beide sind generisch und benötigen die Angabe des Typs bei der Deklarierung. Ein <i>LiveData</i>-Objekt speichert den Wert in seinem Feld <i>value</i>.</p>
<p><i>LiveData</i> werden im <i>ViewModel</i> als öffentliche Felder angegeben. Das erste Feld <code>progressionTodayPercent</code> soll die Prozentzahl enthalten, die angibt inwieweit das Tagesziel an Flüssigkeitszunahme schon erreicht wurde. Dieser Wert soll die Grundlage der <i>Progressbar</i> sein.</p>

<blockquote><pre><code>val progressionTodayPercent = MutableLiveData&#60;Int>(0)</code><pre></blockquote>

<p>Ein zweites Attribut <code>progressionToday</code> soll das absolute Volumen speichern. Implementieren Sie diese selbst.</p>
<p>Bei Übergabe eines Wertes, wird <i>value</i> direkt auf diesen gesetzt. Da das Feld aber <i>nullable</i> ist, muss kein Parameter angegeben werden.<br />Wenn ein Eintrag im <i>InputFragment</i> hinzugefügt wird, soll das Fragment die Methode <code>setProgressionToday</code> des <i>ViewModels</i> mit dem Volumen als Parameter aufrufen. Diese Methode soll die Übergabe an die beiden <i>LiveData</i> weitergeben:</p>

<blockquote><pre><code>fun setProgressionToday (addition: Int){
    progressionToday.value = addition
    progressionTodayPercent.value = (addition/VOLUME_PER_DAY * 100.0).toInt()
 
    if(progressionTodayPercent.value?.compareTo(100) == 1)
        progressionTodayPercent.value = 100
}</code><pre></blockquote>

<p>Die Konstante <code>VOLUME_PER_DAY</code> soll auf <code>1500.0</code> gesetzt werden.<br />
Eine aktuell ausreichende Implementierung des <i>ViewModels</i> ist abgeschlossen. Im letzten Schritt muss das Fragment die beiden Felder beobachten und bei Änderung die <i>Progressionbar</i> anpassen. Fügen Sie dafür in <i>onCreateView</i> von <i>MainFragment</i> hinzu:</p>

<blockquote><pre><code>viewModel.progressionTodayPercent.observe(viewLifecycleOwner, {
    updateProgressBar(it)
})</code><pre></blockquote>

<p>Mit der Methode <i>observe</i> und der Angabe des <i>viewLifecycleOwner</i> kann die <i>LiveData</i> beobachtet werden. Findet eine Änderung statt, wird der Lambda-Ausdruck ausgeführt, der den zweiten Parameter darstellt. Auf die <i>value</i> kann wie gewohnt mit <i>it</i> zugegriffen werden.<br /><p>

<p>Ein kleines Problem muss noch behoben werden. Beide <i>MutableLiveData</i>-Objekte können von jeder Klasse verändert werden, da sie öffentlich sind. Eine Möglichkeit wäre sie direkt als <i>LiveData</i> zu deklarieren. In diesem Fall könnte aber das <i>MainViewModel</i> diese nicht mehr verändern. Deswegen hilft hier ein Trick, der ein bereits kennengelerntes Feature von Kotlin verwendet: benutzerdefinierte <i>Getter</i>. Statt einem, werden zwei Felder deklariert. Ein privates, das eine <i>MutableLiveData</i> beinhaltet und somit nur von der aktuellen Klasse verändert werden kann. Das Zweite, öffentliche, besitzt den Typ <i>LiveData</i> und gibt anstelle des eigenen Wertes den der <i>MutableLiveData</i> zurück. Die Implementierung sieht folgendermaßen aus:</p>

<blockquote><pre><code>val progressionTodayPercent: LiveData&#60;Int>
    get() = _progressionTodayPercent
private val _progressionTodayPercent = MutableLiveData&#60;Int>(0)</code><pre></blockquote>

<p>Wandeln Sie das zweite Feld ebenfalls um und aktualisieren Sie die Referenzen.<br />
    Das absolute Volumen, das in <i>progressionToday</i> gespeichert wird, soll erst einmal in der <i>TextView</i> <code>dataToday_textView</code> angezeigt werden. Fügen Sie <i>MainFragment</i> als Beobachter der <i>LiveData</i> hinzu und implementieren Sie den Lambda-Ausdruck entsprechend.</p>

<h2 style="display:none">Lektion 3 - Datenbank</h2>

<img src="images/Architektur3.png" style="margin-left: 10px; float: right" />

<p>Werden in einer App Daten gespeichert oder aus dem Internet abgerufen, ist eine Datenbank im Hintergrund essentiell. Aus ihr können die Einträge schnell in die App transferiert werden. Auch in der <i>DrinkDrinkDrink!</i>-App sollen die Einträge in einer Datenbank abgelegt werden. Dadurch soll außerdem eine Historie generiert werden, die in der vierten Lektion Verwendung findet.<br />
Das Android-Framework, das SQLite Datenbanken zur Verfügung stellt, heißt <i>Room</i>. Aufgrund der Grundlage SQL sind rudimentäre Fähigkeiten vorteilhaft. Alternativ können Sie sich <a target="_blank" href="https://de.wikipedia.org/wiki/SQL">hier</a> einen kurzen Überblick verschaffen. Das Framework ist in Ihrem Projekt bereits importiert und kann verwendet werden. Eine Besonderheit von <i>Room</i> ist, dass der Zugriff auf die Datenbank asynchron mit Hilfe von <i>Coroutines</i> stattfinden muss. Mit diesen können Aufgaben im Hintergrund ausgeführt werden.<br />
Der Zugriff auf die Datenbank erfolgt in der aktuelle Architektur direkt über das <i>ViewModel</i>. Werden in diese keine Netzwerkdaten eingefügt, kann dies durchaus auch so implementiert werden. Spätestens dann sollte aber eine neue Architekturschicht, ein <i>Repository</i> verwendet werden, das die Datenbank und das Netzwerk verwaltet. Das <i>ViewModel</i> besitzt dabei eine Referenz auf das <i>Repository</i>.</p>

<h3>Modelle</h3>
<p>Da nun mit der Datenbank, der erste <i>externe</i> Bereich hinzukommt, ist ein kurzer Einschub bezüglich der Trennung der Datentypen nötig. Da in jedem Teil verschiedene Arten von Klassen benötigt werden, wird jeweils ein eigener Typ definiert. Mit Erweiterungsmethoden wird dann die Konvertierung zwischen den Datenklassen sichergestellt:</p>

<img src="images/Model.png" />
<br />
<p>Datenbankeinträge werden als Datenklassen repräsentiert und erhalten die Endung <i>Entity</i>. Beim Abfragen der Daten aus dem Internet, werden diese in Datenklassen gespeichert, die das Suffix <i>Property</i> besitzen. Um sie in einer Datenbank zu speichern, muss eine Erweiterungsmethode <i>asDatabaseEntity()</i> bereitgestellt werden. Das <i>Viewmodel</i> hat auch seine eigenen Datenklassen, die <i>DomainModels</i>. Diese besitzen kein Suffix und repräsentieren den Typ, mit dem in den Aktivitäten und Fragmenten gearbeitet wird. Soll also ein Datenbankeintrag dargestellt oder verarbeitet werden, muss dieser zuerst mit <i>asDomainModel()</i> umgewandelt werden. Umgekehrt kann ein <i>DomainModel</i> mit Hilfe von <i>asDatabaseEntity</i> in eine Datenbank eingefügt werden.</p>

<h3>Datenbank</h3>
<p>Voraussetzung für die Implementierung des <i>Repositories</i> ist die Datenbank. Erstellen Sie dafür ein neues Paket <code>database</code>, in das alle zugehörigen Klassen geschrieben werden. Eine Datenbank ist mit <i>Entities</i> gefüllt. Dafür soll eine eigene Datenklasse <i>DrinkDatabaseEntitiy</i> in einer eigenen Datei erstellt werden. Mit der Annontation <code>@Entity</code> wird sie als Datenbankeintrag gekennzeichnet. In der Datenbank soll gespeichert werden, wie viel (<code>volume: Int</code>) wann (<code>time: Date</code>) getrunken wurde. Wie in SQL üblich, muss jeder Eintrag einen primären, einzigartigen Schlüssel, gekennzeichnet mit <code>@PrimaryKey</code> aufweisen. Deswegen sollen Sie für jeden Eintrag in Form einer <code>id</code> automatisch generiert werden:</p>

<blockquote><pre><code>@Entity
data class DrinkDatabaseEntity (
    @PrimaryKey(autoGenerate = true)
    val id: Int,
    val volume: Int,
    val time: Date)</code><pre></blockquote>

<p>Da einer <i>Room</i>-Datenbank nur klassische Datentypen und somit keine Daten (<i>Date</i>) hinzugefügt werden können, muss ein Umweg über den <i>Timestamp</i>, eine numerische Repräsentation des Datums, gegangen werden, der aus einem Datum generiert wird. Dies geschieht mit einem Konvertierer-Objekt:</p>
    
<blockquote><pre><code>class Converters {
    @TypeConverter
    fun fromTimestamp(value: Long?): Date? {
        return value?.let { Date(it) }
    }
 
    @TypeConverter
    fun dateToTimestamp(date: Date?): Long? {
        return date?.time
    }
}</code><pre></blockquote>
    
<p>Der Zugriff auf die Datenbank wird über eine <i>Dao</i>-Schnittstelle ermöglicht. Erstellen Sie dafür ein neues <code>interface DrinkDatabaseDao</code>, das die Annotation <code>@Dao</code> besitzt. In dieser können sqlähnliche Befehle definiert werden. Implementieren Sie die folgenden Methoden:</p>
<ul>
    <li><code>insert</code> besitzt die Annotation <code>@Insert</code> und fügt ein übergebenes Objekt der Datenbank hinzu.</li>
    <li><code>clear</code> entfernt alle Einträge aus der Datenbank. Dazu wird der Sql-Befehl <code>DELETE FROM <i>nameDerDatenbank</i></code> verwendet, der der Annotation <code>@Query</code> übergeben wird. Der Name der Datenbank wird später aktualisiert.</li>
    <li><code>getAll</code> gibt alle Einträge der Datenbank als Liste in einer <i>LiveData</i> zurück. Dafür wird wieder <code>@Query</code> mit dem Befehl <code>SELECT * FROM nameDerDatenbank</code> verwendet. Mit dem Zusatz <code>ORDER BY time DESC</code> werden die Einträge sortiert zurückgegeben. <h3>Exkurs: Koroutinen</h3>
<p>Wie angekündigt wird der Zugriff auf die Datenbank über ein <i>Repository</i> stattfinden. Dieses verwaltet den Zugriff auf die Datenbank und später das Internet. Da diese Operationen einige Zeit in Anspruch nehmen können, werden  beziehungsweise müssen diese asynchron durchgeführt werden. In Kotlin stehen dafür <i>Coroutines</i> bereit, die in diesem Exkurs rudimentär eingeführt werden. Die Anweisungen führt ein neuer Thread aus, der entweder eigenständig oder mit dem Hauptthread interagieren kann. Ein Vorteil ist, dass der Hauptthread durch die Operationen nicht blockiert wird und seine Berechnungen weiterführen kann. Um sie in Notebooks zu verwenden, muss zunächst die zugehörige Bibliothek importiert werden:</p>

<blockquote><pre><code>@file:DependsOn("org.jetbrains.kotlinx:kotlinx-coroutines-core:1.5.0")
import kotlinx.coroutines.*</code><pre></blockquote>

<p>Es folgt ein einfaches Einführungsbeispiel:</p>

```kotlin
@file:DependsOn("org.jetbrains.kotlinx:kotlinx-coroutines-core:1.5.0")
import kotlinx.coroutines.*

GlobalScope.launch { //Startet die Koroutine
    var i = 0
    while(i < 5){
        println("Coroutine $i")
        delay(100L)
        i++
    }
}
println("Hauptthread")
Thread.sleep(2000L) //Hauptthread muss warten, da sonst die Koroutine vorzeitig abgebrochen wird
```

<p>Es wird vom Hauptthread eine Koroutine mit <code>GlobalScope.launch</code> gestartet. In dieser wird in einer Schleife fünf Mal ein Text auf der Konsole ausgegeben. Der Hauptthread <i>schläft</i> zum Schluss kurz, sodass die Koroutine noch beendet wird.<br />
Es sind aber bereits schon einige Fachbegriffe gefallen, die hier näher erklärt werden:</p>
<ul>
    <li><b>Threads</b> werden von der Java Virtual Machine (JVM) verwaltet und verwenden die Multi-Threading-Funktionen des Betriebssystems. Sie besitzen zudem einen eigenen Stack-Speicher. Gemeinsame Informationen werden über einen Heap-Speicher mit anderen Threads geteilt.</li>
    <li><b>Koroutinen</b> werden von Kotlin verwaltet und sind nicht an einen konkreten CPU gebunden, sondern können beliebig verschoben werden. Beispielsweise wird eine Koroutine zuerst auf Kern 1 gestartet und dann auf dem Dritten beendet. Sie sind deutlich flexibler als Threads. Zudem besitzen sie keinen eigenen Speicher und sind ressourcensparsamer. Im Hintergund werden sie auf einer geringen Anzahl an Threads ausgeführt.</li>
    <li>Ein <b>Scope</b> ist der Kontext einer Koroutine. <i>GlobalScope</i> ist beispielsweise die <i>Top-Level</i>-Ebene.</li>
    <li>Der <b>Dispatcher</b> ist mit dem <i>Scope</i> verbunden und entscheidet von welchem Thread die Koroutine ausgeführt wird. Außerdem lässt sich aus diesem der Aufgabenbereich ableiten:</li>
</ul>

```kotlin
import kotlinx.coroutines.*

runBlocking {
    launch { // Besitzt den Kontext des Main-Treads
        println("main runBlocking      : Thread-Name:  ${Thread.currentThread().name}")
    }
    launch(Dispatchers.IO) { // Wird verwendet, wenn im Hintergund auf Dateien oder Netzwerk-Ressourcen zugegriffen wird
        println("IO                    : Thread-Name:  ${Thread.currentThread().name}")
    }
    launch(Dispatchers.Unconfined) { // Verwendet den aktuellen Thread
        println("Unconfined            : Thread-Name:  ${Thread.currentThread().name}")
    }
    launch(Dispatchers.Default) { // Für CPU-intensive Aufgaben
        println("Default               : Thread-Name:  ${Thread.currentThread().name}")
    }
    launch(newSingleThreadContext("eigenerThread")) { // Wird von einem eigenen Thread ausgeführt
        println("newSingleThreadContext: Thread-Name:  ${Thread.currentThread().name}")
    }
}
```

<p>Gestartet können Koroutinen mit dem Schlüsselwort <i>launch</i>, welches üblicherweise auf einem Kontext aufgerufen wird. Zusätzlich kann der <i>Dispatcher</i> bestimmt werden. Bei keiner Angabe, wird der <i>Default-Dispatcher</i> verwendet. Eine weitere Möglichkeit ist das Verwenden von <i>launch</i> in <i>runBlocking</i>, das den aktuellen Thread stoppt und nach Beendigung der Koroutinen wieder aktiviert.</p>

```kotlin
print("Anfang")
runBlocking {
    repeat(500) { //wird 500 mal ausgeführt
        launch { //startet eine Koroutine
            print(".")
        }
    }
}
print("Ende")
```

<p>Mit <i>join</i> kann auf eine gestarteten Koroutine gewartet werden.</p>

```kotlin
import kotlinx.coroutines.*

val job = GlobalScope.launch { //Startet die Koroutine und weist sie der Variable job zu
        delay((500..10000L).random())
        println("Fertig")
}

runBlocking { //Wartet auf die Koroutine
    job.join()
}

println("Koroutine beendet")
```

<p>Soll eine Koroutine ein Ergebnis zurückgeben, findet statt <i>launch async</i> Anwendung:</p>

```kotlin
import kotlinx.coroutines.*

val text = "Kotlin ist Toll"
val laenge = GlobalScope.async(Dispatchers.IO){
    delay(500L) //Wird diese Zeile einkommentiert, wirft getComplete() keinen Fehler
    text.length
}

println("Typ vom laenge: ${laenge::class.simpleName}")

try{
  println("laenge.getCompleted(): ${laenge.getCompleted()}")  
}
catch (e: Exception){
    println("Fehler: $e")
}

runBlocking {
    println("laenge: ${laenge.await()}")
}
```

<p>Die Rückgabe ist vom Typ <i>Deferred&#60;Int></i>. Zusätzlich zum Wert ist gespeichert, ob die Koroutine beendet ist oder nicht. Mit <i>await()</i> in Verbindung mit <i>runBlocking</i> oder <i>getCompleted()</i> kann auf das Ergebnis gewartet werden. Zweiteres löst aber eine <i>IllegalStateException</i> aus, falls die Berechnung noch läuft.</p>

<p>Mit <i>cancel()</i> kann eine <i>Coroutine</i> abgebrochen werden:</p>

```kotlin
import kotlinx.coroutines.*

val job = GlobalScope.launch { //Startet die Koroutine
        delay((500..10000L).random())
        println("Fertig")
}

job.cancel()
println("Koroutine abgebrochen")
```

<p>Bis jetzt wurde der Code der Koroutine immer in einem Lambda-Ausdruck angegeben. Dieser wird immer benötigt, jedoch können die Zeilen in eigene Funktionen geschoben werden. Diese müssen mit dem Schlüsselwort <i>suspend</i> gekennzeichnet sein.</p>

```kotlin
import kotlinx.coroutines.*

suspend fun fak(a: Int): Int{
    var result = 1
    var i = a

    while(i > 1){
        result *= i
        i--
    }
    return result
}

runBlocking{
    val result = async { //Startet die Koroutine
        fak(5)
    }
    println("fak(5): ${result.await()}")
}
```

<h2 style="display:none">Lektion 5 - Standortzugriff</h2>

<p>In der nächsten Lektion wird der Zugriff auf Internetressourcen, eine API, behandelt. Das Ziel ist, dass anhand der Höchsttemperatur am Standort des Geräts die zu empfehlende Trinkmenge angepasst wird. Zuvor lernen Sie aber noch, wie der Standort des Gerätes per GPS bestimmt werden kann.</p>

<p>Es gibt zwei Möglichkeiten den Ort eines Gerätes herauszufinden: per GPS oder über das Netzwerk. Bei ersterem werden Satelliten eingesetzt, die um die Erde kreisen. Letzteres ist schneller, aber ungenauer, da Mobilfunktürme zum Einsatz kommen, die den Standort per Triangulation bestimmen. Weil für diese App eine <i>langsame</i> Bestimmung per GPS ausreicht, wurde diese Methode gewählt.</p>
<p>Bevor aber mit der Programmierung gestartet werden kann, müssen der App zunächst die zugehörigen Freigaben gegeben werden. Würde aktuell versucht werden auf den Standort zuzugreifen, wäre dies nicht möglich, da die Berechtigung der App, fehlt diesen Dienst zu verwenden. Das kann aber in <i>AndroidManifest.xml</i> geändert werden. Fügen Sie vor dem <i>application</i>-Tag folgende Zeilen hinzu:</p>

<blockquote><pre><code>&#60;uses-permission android:name="android.permission.INTERNET" />
&#60;uses-permission android:name="android.permission.ACCESS_FINE_LOCATION" /></code><pre></blockquote>

<p>Zusätzlich zu den Standortfreigaben wird auch gleich der Zugriff auf das Internet erlaubt und somit die darauffolgende Abfrage der Temperatur über eine API. Jetzt kann mit der Implementierung begonnen werden.<br />
Eine wichtige Frage ist, in welchem Teil der App die Standortbestimmung ausgeführt werden soll. Der Fortschritt wird im <i>MainFragment</i> angezeigt und im zugehörigen <i>ViewModel</i> berechnet. Auf den ersten Blick scheint also dort der richtige Ort dafür zu sein. Jedoch würde bei einem Fragmentwechsel die Ausführung gestoppt werden. Dann bleibt nur noch die Aktivität. Diese wird nicht gewechselt, da es nur eine gibt, und alle Fragmente können auf sie zugreifen. Dadurch kann in diesem, sehr vereinfachten, Fall eine Datenbank gespart und die Daten direkt weiterverarbeitet werden.<br />
Die Standortabfrage ist bereits in Android vorhanden und muss nur noch der App hinzugefügt werden. Erstellen Sie dafür ein privates, später initialisiertes Feld <code>locationManager</code> vom Typ <code>LocationManager</code>. In der Methode <i>onCreate</i> wird dieses dann mit</p>

<blockquote><pre><code>locationManager = getSystemService(Context.LOCATION_SERVICE) as LocationManager</code><pre></blockquote>

<p>befüllt. Anhand des Codes wird nochmal klar, dass bereits eine systemweite Standortsuche vorhanden ist, auf die zugegriffen wird. Die eigentliche Bestimmung soll in eine private Methode ausgelagert werden. Nennen Sie diese <code>calculateLocation</code>. <br />
Egal welches Betriebssystem Sie auf Ihrem Handy besitzen, bei der Installation einer App, die den Standort benötigt, werden Sie immer gefragt, ob sie dies erlauben möchten. Wenn Sie aber den aktuellen Zustand der <i>DrinkDrinkDrink!</i>-App ausprobieren, passiert nichts. Sie haben jedoch nicht das System ausgetrickst, sondern es fehlt die Implementierung der Abfrage. Vor jeder GPS-Abfrage muss überprüft werden, ob die Berechtigung erteilt wurde und wenn nicht, muss sie eingeholt werden:</p>

<blockquote><pre><code>if (ActivityCompat.checkSelfPermission(
        applicationContext,
        Manifest.permission.ACCESS_FINE_LOCATION
    ) != PackageManager.PERMISSION_GRANTED
) {
    ActivityCompat.requestPermissions(
        this,
        arrayOf(
            Manifest.permission.ACCESS_FINE_LOCATION
        ),
        101
    )
}</code><pre></blockquote>

<p>Wenn Sie nun die Methode an eine passende Stelle des <i>Lifecycles</i> einfügen, sollten Sie nun beim Start der App nach der Berechtigung gefragt werden.<br />
    Jetzt kann der eigentliche Code in den <i>else</i>-Block der <i>if</i>-Verzweigung eingefügt werden:</p>

<blockquote><pre><code>locationManager.requestLocationUpdates(LocationManager.GPS_PROVIDER, 60000, 1000F, LocationListener {
    it.let {
        val longitude = it.longitude
        val latitude = it.latitude
        Timber.i("longitude: $longitude, latitude: $latitude")
    }
})</code><pre></blockquote>

<p><i>requestLocationUpdates</i> fragt in regelmäßigen Abständen den Standort mit der angegeben Methode, hier GPS, ab. Die Zeit zwischen den Abfragen beträgt hier 60000ms, also 1 Minute. Zudem kann auch eine Meldung ausgelöst werden, wenn eine bestimmte Distanz zurückgelegt wurde. Diese wurde hier auf 1000 Meter gesetzt. Diese Werte können Sie aber auch nach Belieben ändern. Wichtig ist der letzte Parameter, der <i>LocationListener</i>. Dieser legt fest, wie nach einer Abfrage verfahren wird. Dem zugehörigen Lambda-Ausdruck wird dabei der Standort übergeben. In der Vorlage wird der Längen- und Breitengrad in jeweils einer Variable gespeichert und im Log ausgegeben.<br />
Wenn Sie den Emulator verwenden, können Sie das Gerät schlecht bewegen. Deswegen kann dies ebenfalls simuliert werden. In dem Menü, welches über die drei Punkte in der Seitenleiste erreicht werden kann, gibt es die Möglichkeit unter dem Punkt <i>Location</i> Standorte zu bestimmen. Wählen Sie einen in der Karte aus, speichern und setzen Sie diesen, indem Sie ihn in der Liste auswählen und <i>SET LOCATION</i> betätigen. Im Log sollten dann neue Koordinaten angezeigt werden.</p>

<h2 style="display:none">Lektion 6 - Netzwerkzugriff</h2>
<p>Die aktuelle Temperatur an dem Standort soll mit Hilfe einer Wetter-Api erhalten werden. Falls Sie noch nicht mit REST-Apis in Berührung kamen, gibt Ihnen <a target="_blank" href="https://towardsdatascience.com/introduction-to-rest-apis-90b5d9676004">dieser Artikel</a> sowie die <a target="_blank" href="https://de.wikipedia.org/wiki/Representational_State_Transfer">zugehörige Wikipedia-Seite</a> einen kurzen Überblick. Für diesen Kurs wurde <a target="_blank" href="http://www.7timer.info/doc.php?lang=en">7Timer!</a> gewählt, da die Daten den Anforderungen entsprechen, sie gut strukturiert ausgegeben werden und kein Account notwendig ist. Abgefragt werden die Wetterdaten über eine neue Bibliothek <a target="_blank" href="https://square.github.io/retrofit/">retrofit2</a>. Sie überzeugt durch ihre einfache Implementierung und einem Aufbau, der einer <i>Room</i>-Datenbank ähnelt. Rückgabe der Anfrage wird eine JSON-Datei sein, welche mit Hilfe von <a target="_blank" href="https://github.com/square/moshi/">moshi</a> in Objekte von Datenklassen konvertiert wird. Es wurden im Grundgerüst bereits alle Bibliotheken hinzugefügt. Sie können also direkt von Ihnen verwendet werden.</p>

<p>Erstellen Sie ein neues Paket <code>network</code> in <i>com.example.android.drinkdrinkdrink</i>. Diesem werden im Laufe der Lektion zwei Dateien hinzugefügt. Die Erste, um die Sie sich zuerst kümmern werden, beinhaltet den <i>retrofit</i>-Dienst, sowie eine Schnittstelle, in der alle Abfragen definiert sind. In der zweiten Datei werden Sie daraufhin die Netzwerkmodelle definieren, in die die Rückgabe der Abfrage mit <i>moshi</i> konvertiert wird.<br />
Aber zuerst muss <i>retrofit</i> konfiguriert werden. Wie bei der Datenbank, ist auch hier das Herz eine Schnittstelle. Erstellen Sie für diese eine Datei und nennen Sie sie <code>WeatherApiService</code>. In dieser können unter Anderem <i>GET</i> und <i>POST</i> Anfragen an eine Webseite geschickt werden. Lassen Sie sie aber zunächst leer.<br />
Um die Anfragen absenden zu können, ist ein Objekt der Klasse <i>Retrofit</i> notwendig. Dieses wird üblicherweise in einem öffentlichen Objekt gespeichert, über das auf den Dienst zugegriffen werden kann. Zusätzlich muss der JSON-Konvertierer erzeugt und dem Dienst hinterlegt werden. Erstellen Sie dazu dem Objekt <code>Network</code>. Dadurch wird, wie bei der Datenbank, das Entwurfsmuster Singleton verwendet. In das Objekt soll zunächst das private Feld <code>moshi</code> hinzugefügt werden:</p>

<blockquote><pre><code>private val moshi = Moshi.Builder()
    .add(KotlinJsonAdapterFactory())
    .build()</code><pre></blockquote>

<p>Bevor der Retrofit-Dienst erstellt werden kann, muss die Ziel-Url angegeben werden. Dafür bietet sich eine private Konstante an, die <code>https://www.7timer.info/bin/</code> als Zeichenkette speichert. Fügen Sie diese dem Anfang der Datei hinzu.</p>
<details>
    <summary>Lösung</summary>
<blockquote><pre><code>private const val BASE_URL = "https://www.7timer.info/bin/"</code><pre></blockquote>
</details>

<p>Mit der Ziel-Url und dem <i>moshi</i>-Konvertierer kann der <i>retrofit</i>-Dienst erstellt werden. Die ersten beiden Aufrufe auf den <a target="_blank" href="https://de.wikipedia.org/wiki/Erbauer_(Entwurfsmuster)">Erbauer</a> verarbeiten genau diese Informationen:</p>

<p>In die Schnittstelle <i>WeatherApiService</i> muss nun der passende Aufruf eingefügt werden. Da die Dauer der Antwort variieren und manchmal sogar wenige Sekunden dauern kann, ist es sinnvoll diese Aufgabe einer Koroutine zuzuordnen. Deswegen muss der Methode <code>getDataAsync</code> das Schlüsselwort <code>suspend</code> vorangestellt werden. Das Konzept des Aufrufs ist ähnlich zu einem Datenbankbefehl. <blockquote><pre><code>@GET("civillight.php")

In der <a target="_blank" href="http://www.7timer.info/doc.php?lang=en#variables">Dokumentation</a> können Sie die verfügbaren Variablen einsehen. <h2 style="display:none">Lektion 7 - RecyclerView</h2>

<img src="images/recyclerView.gif" style="float: right; margin-left: 10px"/>

<p>Mit der <i>RecyclerView</i> wird ein wichtiges Feature in der App-Entwicklung bereitgestellt: Soll eine beliebig lange Liste an Objekten dargestellt werden, war dies nur mit einer <i>SrollView</i> möglich. Dieser könnte für jedes Element der Liste ein eigenes Layout angefügt werden. Dieses Vorgehen hat aber zwei Nachteile: die aufwändige Programmierung und der Ressourcenverbrauch. Alle Einträge werden zu Beginn erstellt, unabhängig davon, ob sie zu sehen sind. Genau hier setzt die <i>RecyclerView</i> an. Es werden zu Beginn nur die Einträge generiert, die auch wirklich zu sehen sind. Wird dann ein Element der Liste aus dem Sichtfeld bewegt (zum Beispiel nach oben), wird das verwendete Layout wieder unten mit neuen Daten angefügt. Es wird wiederverwendet. Zudem überzeugt das Konzept auch mit einer einfachen Implementierung, die im Folgenden beleuchtet wird.<br />
In der <i>DrinkDrinkDrink!</i>-App soll eine <i>RecyclerView</i> im <i>HistoryFragment</i> verwendet werden, um alle gespeicherten Einträge ansprechend darzustellen.</p>

<p>Die Daten, genauer alle Datenbankeinträge, müssen aufgrund der Architektur wieder in einem <i>ViewModel</i> verwaltet werden. Erstellen Sie ein neues Paket <code>history</code> in <i>fragments</i>, verschieben Sie das <i>HistoryFragment</i> in dieses und erstellen Sie ein <code>HistoryViewModel</code>, sowie eine <code>HistoryViewModelFactory</code> nach dem Vorbild des <i>MainFragments</i>. Das <i>HistoryViewModel</i> soll ein öffentliches Feld besitzen, das alle Datenbankeinträge der <i>DrinkDatabase</i> beinhaltet.</p>

<details>
    <summary>Lösung <i>HistoryViewModel</i></summary>
<blockquote><pre><code>class HistoryViewModel(application: Application): ViewModel() {
 
    private val drinkDatabase: DrinkDatabase = getDrinkDatabase(application)
    private val drinkRepository: DrinkRepository = DrinkRepository(drinkDatabase)
 
    val drinks = drinkRepository.drinks
 
}</code><pre></blockquote>
</details>

<details>
    <summary>Lösung <i>HistoryViewModelFactory</i></summary>

<h3>Adapter</h3>
<p>Das Herzstück der Implementierung ist ein <a target="_blank" href="https://de.wikipedia.org/wiki/Adapter_(Entwurfsmuster)">Adapter</a>. Erstellen Sie für diesen eine neue Klasse <code>HistoryAdapter</code>. Sie soll von <code>RecyclerView.Adapter&#60;HistoryAdapter.ViewHolder>()</code> erben.<br />
Zuerst muss in dem Adapter ein öffentliches Feld angelegt werden, in dem die Daten verwaltet werden. Nennen Sie es <code>data</code>. Zu Beginn soll es mit einer leeren Liste von <code>Drink</code>s gefüllt werden. Definieren Sie zusätzlich einen eigenen <i>Setter</i>, der zusätzlich zur normalen Zuweisung die Methode <code>notifyDataSetChanged()</code> aufruft. Mit dieser wird der <i>RecyclerView</i> mitgeteilt, dass die Datengrundlage verändert wurde und die Elemente neu erstellt werden müssen.</p>

<details>
    <summary>Lösung</summary>
<blockquote><pre><code>var data = listOf&#60;Drink>()
    set(value) {
        field = value
        notifyDataSetChanged()
    }</code><pre></blockquote>
</details>

<p>Überschreiben Sie daraufhin die Methode <code>getItemCount()</code>. Diese soll die Anzahl der Einträge in <i>data</i> zurückgeben. Durch diese erfährt die <i>RecyclerView</i> wie viele Elemente Sie enthalten muss.</p>

<details>
    <summary>Lösung</summary>
<blockquote><pre><code>override fun getItemCount() = data.size</code><pre></blockquote>
</details>

<p>Jeder Eintrag besitzt per Vorgabe einen eigenen <i>ViewHolder</i>, der noch implementiert werden muss. Die Verbindungen vom <i>HistoryAdapter</i> und <i>ViewHolder</i> können aber schon vorbereitet werden. <code>onCreateViewHolder()</code> erstellt einen <i>ViewHolder</i> und gibt diesen zurück:</p>

<blockquote><pre><code>override fun onCreateViewHolder(parent: ViewGroup, viewType: Int): ViewHolder{
    val layoutInflater = LayoutInflater.from(parent.context)
    val binding = ListItemDrinkBinding.inflate(layoutInflater, parent, false)
    return ViewHolder(binding)
}</code><pre></blockquote>

<p>Nach der Erstellung müssen die Daten noch mit dem Layout verbunden werden. Dies geschieht in <code>onBindViewHolder</code>:</p>
<blockquote><pre><code>override fun onBindViewHolder(holder: ViewHolder, position: Int) {
    val item = data[position]
    holder.bind(item)
}</code><pre></blockquote>

<p>Diese beiden Methoden sind essentiell für das Recyceln der Layouts. Zu Beginn werden mit <i>onCreateViewHolder</i> so viele Instanzen erstellt, wie dargestellt werden können. Danach erfolgt eine Füllung dieser mit den ersten Elementen durch <i>onBindViewHolder</i>. Wird dann die Ansicht verschoben, werden lediglich neue Daten in die <i>ViewHolder gebunden</i>. Die <i>RecyclerView</i> merkt sich dabei welche Elemente das sein müssen.</p>

<h3>ViewHolder</h3>
<p>Der <i>ViewHolder</i>, der ein Layout der <i>RecyclerView</i> und einen Eintrag repräsentiert, wird als innere Klasse realisiert. Sie erbt von <code>RecyclerView.ViewHolder</code> und bekommt bei Erzeugung ein <code>ListItemDrinkBinding</code>-Objekt übergeben. Der Super-Klasse muss das äußerste Layout des Parameters übergeben werden.<br />
Die einzige Funktionalität, die bereitgestellt werden muss, ist die Darstellung eines Datenelements in dem Layout. Dies soll die Methode <code>bind</code> bereitstellen, die ein Objekt der Klasse <code>Drink</code> übergeben bekommen soll. Das Layout besitzt nur eine einzige <i>TextView</i>, die mit einem formatierten Text gefüllt werden soll. Bei der Formatierung des Datums können Sie sich an der Methode <i>transformDrinksToday</i> der Klasse <i>MainViewModel</i> orientieren. Die entsprechenden Formatierungsmuster können Sie <a target="_blank" href="https://docs.oracle.com/javase/8/docs/api/java/text/SimpleDateFormat.html">hier</a> nachschlagen.</p>

<details>
    <summary>Lösung</summary>
<blockquote><pre><code>class ViewHolder(private val binding: ListItemDrinkBinding): RecyclerView.ViewHolder(binding.root) {
    fun bind(item: Drink) {
        val local = Locale("de")
        val fmtDatum = SimpleDateFormat("dd.MM.yyyy", local)
        val fmtZeit = SimpleDateFormat("HH:mm", local)
 
        binding.dataTextView.text = "${item.volume}ml am ${fmtDatum.format(item.time)} um ${fmtZeit.format(item.time)} Uhr"
    }
}</code><pre></blockquote>
</details>

binding.historyRecyclerView.adapter = historyDrinksAdapter

</details>

<h2 style="display:none">Lektion 8 - Veröffentlichung von Apps</h2>

<p>In dieser kurzen, abschließenden Lektion wird der Prozess beschrieben, wie eine App im <i>Google Play Store</i> veröffentlicht werden kann. Als Erstes ist ein Entwicklerkonto für die <a target="_blank" href="https://play.google.com/apps/publish">Google Play Console</a> Voraussetzung. Dafür können Sie ein bereits bestehendes Google Konto verwenden oder ein Neues erstellen. Während der Registrierung wird eine einmalige Gebühr von 25$ fällig, die Sie zum Beispiel per Kreditkarte begleichen müssen. Daraufhin können Sie ihr Entwicklerkonto in den Menüpunkten <i>Kontodetails</i> und <i>Entwicklerseite</i> mit Informationen befüllen. Letzterer Punkt ist wichtig für Ihre Darstellung im Store.<br />
Eine Übersicht über Ihre Apps bekommen Sie im <i>Dashboard</i>. Eine neue App erstellen, können Sie mit dem zugehörigen Button, rechts oben, mit der Eingabe eines Namens und den Eigenschaften. Danach können Sie in den App-Bereich wechseln. In diesem dreht sich alles um die ausgewählte App. Den Store-Eintrag können Sie im gleichnamigen Menüpunkt bearbeiten. Dieser muss vor dem Veröffentlichen der App vollständig sein. Er enthält alle üblichen Punkte: Name, Beschreibungen, App-Symbol und Screenshots.<br />
Das Wichtigste sind aber die <i>Releases</i>. Das sind <i>App-Bundles</i> oder <i>APKs</i> (Android Package), die Sie in <i>Android Studio</i> generieren können. Jedes Release stellt eine Version der App dar. Diese besteht aus einer eindeutigen Versionsnummer, sowie einer Kennung, die beliebig gewählt werden kann. Beides muss in <i>Android Studio</i> in <i>build.gradle</i> auf Projekt-Ebene definiert werden. Sie sind unter den Variablen <i>versionCode</i> und <i>versionName</i> zu finden. Ersterer muss bei jedem neuen Release um eins erhöht werden.<br />
Ein Release benötigt ein signiertes <i>App-Bundle</i> oder eine signierte <i>APK</i>. Dieses können Sie in <i>Android Studio</i> unter <i>Build</i> -> <i>Generate Signed Bundle/APK</i> erstellen. Wählen Sie <i>Android App Bundle</i> aus. Daraufhin wird ein Schlüssel benötigt, mit dem die Datei signiert wird. Generieren Sie diesen, indem Sie auf <i>Create New</i> klicken. Daraufhin müssen Sie einige Daten eingeben. Wählen Sie einen Speicherplatz, generieren Sie Passwörter und speichern Sie sich diese. Hilfe für die weiteren Felder finden Sie <a target="blank" href="https://developer.android.com/studio/publish/app-signing#generate-key">hier</a>. Daraufhin können Sie den Schlüssel auswählen und das <i>Bundle</i> in im Ordner <i>release</i> generieren.<br />
Sie haben verschiedene Möglichkeiten die App zu veröffentlichen: Intern, Alpha, Beta, Öffentlich.</p>

<ul>
    <li>Intern: Dieser Test ist besonders für Entwickler:innen interessant. Wird ein <i>Release</i> veröffentlicht, kann dieser direkt über einen Link heruntergeladen werden. Ein Update über den <i>Play Store</i> ist nicht möglich. Es ist zudem eine Liste an Tester:innen zu definieren.</li>
    <li>Alpha: In der Konsole wird diese Möglichkeit <i>Geschlossener Test</i> genannt. Der Release kann über den Play Store von einer definierten Gruppe an Tester:innen heruntergeladen und auch dort aktualisiert werden. Jede Version wird jedoch vorher von Google überprüft, was einige Zeit in Anspruch nehmen kann.</li>
    <li>Beta: Für einen Beta oder auch <i>Offenen Test</i> kann sich jede:r Play Store-Nutzer:in anmelden.</li>
    <li>Öffentlich: Geht ein <i>Release</i> in <i>Produktion</i>, ist er für alle im <i>Play Store</i> und kann heruntergeladen werden. Vorher wird er jedoch abermals von Google überprüft.</li>
</ul>

<p>Eine Version kann in verschiedenen <i>Tracks</i> gleichzeitig verwendet und hochgestuft werden. Zum Beispiel kann die Version 21 zunächst als Beta veröffentlicht und wenn keine Probleme auftreten in die Produktion hochgestuft werden. Bevor Sie aber die App veröffentlichen und dem <i>Produktions-Track</i> zuweisen können, müssen Sie noch einige zusätzliche Informationen hinterlegen, die ebenfalls überprüft werden. Eine Übersicht finden Sie <a target="_blank" href="https://support.google.com/googleplay/android-developer/answer/9859455?hl=de&ref_topic=7072031">hier</a>.</p>
