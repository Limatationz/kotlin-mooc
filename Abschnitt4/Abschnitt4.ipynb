<ul>

    <li>Empfohlen: Jetbrains stellt mit seiner Toolbox eine einfache Möglichkeit bereit ausgewählte IDEs zu verwalten. Die Toolbox können Sie sich <a target="_blank" href="https://www.jetbrains.com/de-de/toolbox-app/">hier</a> herunterladen.</li>

<h3>Einrichtung von Android Studio</h3>
<p>Öffnen Sie Android Studio (bei Erstellung des Kurses mit Version 4.2.2) und gehen Sie in die Einstellungen. Um die Apps auf dem Gerät direkt testen zu können, müssen Sie ein <i>Software Development Kit</i> (SDK) herunterladen. Navigieren Sie dafür nach <i>Appearance & Behavior</i> → <i>System Settings</i> → <i>Android SDK</i>. In dem ersten Reiter <i>SDK Platforms</i> werden Ihnen verschiedene Android Versionen angeboten. Zusätzlich zu ihrem Namen wird auch ein API-Level angegeben. Auf dieses werden wir später noch einmal zurückkommen. Laden Sie entweder die Neuste oder Android 11.0 herunter. Falls Sie im Laufe der Installation auf eine Fehlermeldung <i>Failed to install Intel HAXM</i> gestoßen sind, empfehle ich ein weiteres <i>SDK Tool</i> zu installieren. Gehen Sie dafür im gleichen Menüpunkt in den gleichnamigen Bereich und wählen Sie den <i>HAXM installer</i> aus. Dieser verbessert die Performance des Emulators, der im nächsten Schritt eingerichtet wird.</p>

<h3>Einrichtung des Emulators</h3>
<p>Um die Apps direkt auf Ihrem Gerät testen zu können, ist ein Emulator notwendig. Diesen können Sie im <i>AVD Manager</i> einrichten. Beim Starten von Android Studio erreichen Sie diesen im Punkt <i>Configure</i> des sich öffnenden Fensters. Ein neues Gerät können Sie mit dem <i>+</i> hinzufügen. Wählen Sie ein Smartphone aus (zum Beispiel das Pixel 4). Mit dem Drücken von <i>Next</i> kommen Sie zu einer Auswahl des Betriebsystems. Auch hier können Sie sich für eine beliebige Version entscheiden. Die Apps, die in diesem Kurs vorgestellt werden, benötigen mindestens das API-Level 26. Mit der neusten Android-Version sind Sie für diesen Kurs auf der sicheren Seite. Wenn der Download abgeschlossen ist, kommen Sie mit <i>Next</i> in ein Übersichtsfenster. Wahlweise können Sie dem Gerät noch einen besonderen Namen geben. Mit <i>Finish</i> beenden Sie die Einrichtung und das eben konfigurierte Gerät sollte in dem Manager angezeigt werden.</p>

<h3>Das eigene Smartphone als Emulator verwenden</h3>
<p>Falls Sie ein Smartphone besitzen, das Android verwendet, können Sie die Apps auch auf diesem direkt aus Android Studio ausführen. Dazu müssen Sie zunächst auf den Gerät den Entwicklermodus aktivieren. Öffnen Sie die <i>Einstellungen</i> und navigieren dann in das Menü <i>Über das Telefon</i>, das ganz unten zu finden ist. Ebenfalls am Ende der Liste finden Sie den Punkt <i>Build-Nummer</i>. Tippen Sie auf diesen sieben Mal, bis sich der Entwicklermodus aktiviert. Jetzt muss nur noch das <i>USB/Android-Debugging</i> aktiviert werden, das unter <i>System</i> → <i>Entwickleroptionen</i> zu finden ist. Schließen Sie das Smartphone per USB-Kabel an ihr Gerät an. Bei Linux ist eine separate Installation von <i>adb</i> notwendig (<code>sudo apt install abd</code>). Falls Ihr Gerät später nicht erkannt wird, finden Sie <a target="_blank" href="https://developer.android.com/studio/run/device.html">hier</a> Hilfe.</p>

<h2 style="display:none">Lektion 1 - Hello World #2</h2>

<p>Der Abschnitt der App-Entwicklung soll ebenfalls mit <i>Hello World</i> eingeleitet werden. Jetzt aber in Form einer Android-App. Starten Sie Android Studio und erstellen Sie ein neues Projekt. Es öffnet sich ein Fenster in dem Ihnen verschiedene Vorlagen, die bestimmte Funktionen bereits bereitstellen, angeboten werden. Wählen Sie im Punkt <i>Phone and Tablet Empty Activity</i> aus. Im Folgenden sollen Sie einen Namen und einen Paketnamen angeben. Der Name der App soll in Ihrem Fall <i>HelloWorld</i> sein. In der Regel beginnt dieser mit einem Großbuchstaben. Der Paketname besteht aus einem Namensraum, zum Beispiel Ihrer Domain und der vorher definierte Namen der App. Wichtig ist, dass die Domain von hinten nach vorne angegeben wird. Aus <i>uni-bayreuth.de</i> würde <i>de.uni-bayreuth</i> werden. Geben Sie für diese App den Paketnamen <i>com.example.android.helloworld</i> an. Den zugehörigen Namensraum, <i>com.example.android</i>, werden Sie auch bei allen weiteren Projekten dieses Kurses verwenden). Bei der Sprachauswahl sollte Kotlin ausgewählt sein. Als letzten Schritt möchte Android Studio das <i>Api-Level</i> wissen, für die die App erstellt werden soll. Jede neue Version bietet erweiterte Funktionen, jedoch ist für eine große Abdeckung an Geräten die neuste Version unvorteilhaft. Unter <i>Help me choose</i> kann eine Übersicht geöffnet werden, die bei der Entscheidung helfen soll. Für die <i>HelloWorld</i> und alle weiteren Apps ist <i>Android 8.0 (Oreo)</i> ausreichend. Wenn Sie alle Felder richtig ausgefüllt haben, kann das Projekt mit <i>Finish</i> erstellt werden. Bis die Entwicklerumgebung geladen hat, kann es kurz dauern.  
Die Benutzeroberfläche ähnelt der von <i>IntelliJ</i>. Oben ist eine Leiste mit vielen verschiedenen Symbolen zu finden und an den anderen drei Rändern des Fensters sind verschiedene Teilfenster versteckt, die bei Klick auf den Namen eingeblendet werden können. Betrachten Sie zunächst die Projektansicht, die links zu finden ist. Unter dem aktuellen Pfad, kann das Aussehen der Ansicht verändert werden. Falls <i>Android</i> noch nicht ausgewählt ist, ändern Sie diese Einstellung. Klappen Sie alle Ordner auf, sodass Sie den Aufbau des Projekts überblicken können.</p>

<img src="images/OverviewGUI.png" />

<p>Es folgt ein kurzer Überblick über die Ordnerstruktur. Die zugehörigen Dateien werden im Laufe der Abschnitte verändert und dabei genauer kennengelernt.</p>

<ul>
    <li>app: In diesem Modul befinden sich alle Kotlin-Dateien und Ressourcen.</li>
        <ul>
            <li>manifests: Enthält eine Datei, in der alle Eigenschaften der App zu finden sind. Deren Name, der Paketname, sowie alle <i>Ansichten</i>. Zudem sind hier auch die benötigten Berechtigungen (zum Beispiel Internetzugriff) angegeben.</li>
            <li>java: Beinhaltet alle Kotlin-Dateien. Der Name hat eine historische Bedeutung. Er wurde noch nicht geändert, da vor Kotlin Java die empfohlene Programmiersprache war und eine Rückwärtskompatibilität gewährleistet werden soll.</li>
                <ul>
                    <li>com.example.android.helloworld: Dies ist der Ordner, in dem die gesamte Logik der App angesiedelt ist.</li>
                    <li>com.example.android.helloworld (androidTest): Enthält JUnit-Tests, die eine Emulation der App benötigen.</li>
                    <li>com.example.android.helloworld (test): Ebenfalls ein Paket, das JUnit-Tests enthält. Diese können aber ohne Emulation auf einem Android Gerät ausgeführt werden.</li>
                </ul>
            <li>java (generated): Enthält den automatisch generierten Code.</li>
            <li>res: Wie das Symbol schon andeutet, sind hier alle Ressourcen in passenden Ordnern angesiedelt. Dieser Ordner wird im Laufe der beiden Abschnitte um neue Unterordner ergänzt werden.</li>
                <ul>
                    <li>drawable: Bilder, die in der App Verwendung finden</li>
                    <li>layout: In diesem Ordner ist der Aufbau der verschiedenen Ansichten enthalten.</li>
                    <li>mipmap: Beinhaltet das App-Icon.</li>
                    <li>values: Bietet Platz für Ressourcen-Dateien, die Konstanten enthalten. Beispielsweise werden in <i>colors.xml</i> alle Farben gebündelt verwaltet. <i>strings.xml</i> hingegen bietet Platz für Zeichenketten.</li>
                </ul>
        </ul>
    <li>Gradle Scripts: Gradle ist das Build-Tool, mit dem der Kotlin-Code kompiliert und in eine App umgewandelt wird.</li>
        <ul>
            <li>build.gradle (Project): Enthält die Konfiguration für das gesamte Projekt und die Version von Gradle.</li>
            <li>build.gradle (Module): In dieser Datei werden unter Anderem alle Bibliotheken hinzugefügt, die die App verwendet (<i>dependencies</i>). Zudem ist die SDK-Version und auch die Version der App angegeben.</li>
        </ul>
</ul>
  
<p>Ohne auch nur eine Datei zu verändern, kann das Projekt schon ausgeführt werden. Wählen Sie dafür den richtigen Emulator aus (Beispielsweise das eingerichtete Pixel 4) und drücken Sie das <i>Run</i>-Icon. Es öffnet sich nach einiger Zeit der Emulator in einem neuen Fenster. Außerdem wird die App aus dem vorhanden Code von <i>Gradle</i> konstruiert. Die <i>Hello World-App</i> öffnet sich automatisch auf dem virtuellen Gerät.</p>

<h2 style="display:none">Lektion 2 - Log und Debug</h2>

<p>Ein Log ist ein mächtiges Werkzeug in der Programmierung. Stürzt die App ab, kann meistens anhand von den Logeinträgen das Problem gefunden werden. Noch wichtiger ist aber die Möglichkeit des Debuggings, dem schrittweise Ausführen von Code. Beides wird in dieser Lektion behandelt.</p>
<h3>Logging</h3>
<img src="images/ImportTimber.gif" style="float: right; margin-left: 10px"/>

<p>Android bietet selbst eine <a href="https://developer.android.com/reference/android/util/Log">API</a> für diesen Zweck an. Da diese aber im Vergleich zu der im Folgenden vorgestellten Bibliothek anfangs überwältigend wirkt, wird <a href="https://github.com/JakeWharton/timber">Timber</a> verwendet. Nebenbei lernen Sie, wie Bibliotheken in Ihre Apps eingebaut werden können und was eine <i>Application</i> ist.</p>


<h4>Timber als Bibliothek hinzufügen</h4>

<p>Wie bereits erwähnt, sind alle Bibliotheken in <i>Gradle</i> zu finden. Expandieren Sie dazu den <i>Gradle Script</i>-Bereich und öffnen Sie die <i>build.gradle</i>-Datei des Moduls (<i>Module: HelloWorld.app</i>). Ganz unten befindet sich der Abschnitt <i>dependencies</i>. Fügen Sie dem Block als neue Zeile diesen Code hinzu:</p>

<blockquote><code>implementation 'com.jakewharton.timber:timber:4.7.1'</code></blockquote>

<p>Daraufhin erscheint über den Editor eine blaue Zeile, die darauf hinweist, dass eine wichtige Datei verändert wurde und eine erneute Synchronisierung notwendig ist. Ein solcher Vorgang wird immer benötigt, wenn eine der beiden <i>build.gradle</i>-Dateien verändert wird. Bestätigen Sie die Synchronisierung mit <i>Sync Now</i>. Die Bibliothek wurde somit zum Projekt hinzugefügt und kann verwendet werden.<br />
Um <i>Timber</i> in der gesamten App verwenden zu können, muss ein <i>Baum</i> in einer <i>Application-Class gepflanzt</i> werden.</p>

<div style="clear: right" />

<img src="images/Application.gif" style="float: right; margin-left: 10px"/>

<p>Dafür muss eine neue Klasse erstellt werden, die von <i>Application</i> erbt. Klicken Sie mit einem Rechtsklick auf das Paket <i>com.example.android.helloworld</i>, in dem sich bereits die Klasse <i>MainActivity</i> befindet. Wählen Sie über <i>New Kotlin Class/File</i> aus. Der Name für die neue Datei soll <i>MainApplication</i> lauten und die Art <i>Class</i>. Android Studio legt daraufhin die Datei an und füllt sie auch schon mit einer Klasse <i>MainApplication</i>. Diese soll von <i>Application</i> erben. Der Name der Super-Klasse ist aber noch rot hinterlegt. Mit Hilfe eines Rechtsklicks auf die Klasse und der Auswahlmöglichkeit <i>Show Context Actions</i>, können Sie dieses Problem einfach lösen. Wählen Sie <i>Import</i>. Dadurch wird die zugehörige Klasse aus der Android-Bibliothek dem Namensraum hinzugefügt. Die Instanz, der <i>Baum</i>, von Timber soll beim Start der App erzeugt werden. Überschreiben Sie dafür die Methode <i>onCreate()</i>, die beim Öffnen der App aufgerufen wird. Mit dem Kurzbefehl <i>Strg+O</i> können Sie ein Hilfefenster erreichen, das alle Methoden aufzeigt, wie zum Beispiel <i>onCreate()</i>. Sie können dort auch nach bestimmten Methoden suchen. Zuletzt müssen Sie die Auswahl mit <i>Enter</i> bestätigen. Daraufhin wird die Methode der Klasse hinzugefügt. Ergänzen Sie sie mit folgender Zeile:</p>

<blockquote><code>Timber.plant(Timber.DebugTree())</code></blockquote>

<p>Sollte <i>Timber</i> rot hinterlegt sein, müssen Sie die Klasse importieren. Die App weiß jetzt aber noch nicht, dass Sie eine eigene <i>Application</i>-Klasse verwenden möchten. Fügen Sie dazu der Datei <i>AndroidManifest.xml</i> in <i>manifests</i> dem <i>application</i>-Tag ein weiteres Attribut zu:</p>
<blockquote><code>android:name=".MainApplication"</code></blockquote>

<p>Jetzt können Sie in der App einfache Nachrichten dem Log hinzufügen. Probieren Sie es gleich aus! Fügen Sie der Methoden <i>onCreate()</i> der Klasse <i>MainApplication</i> die Zeile:</p>

<blockquote><code>Timber.i("MainApplication created!")</code></blockquote>

<p>und der <i>onCreate()</i> der bereits bestehenden Klasse <i>MainActivity</i></p>

<blockquote><code>Timber.i("MainActivity created!")</code></blockquote>

<p>hinzu.</p>

<div style="clear: right" />

<img src="images/Log.gif" style="float: right; margin-left: 10px"/>

In der oberen Leiste des neuen Fensters sollte links das Gerät, auf dem die App läuft, automatisch ausgewählt werden. Wählen Sie sie ansonsten aus. Rechts davon ist die App zu finden. Wählen Sie als <i>Log-Level Info</i> aus und suchen Sie in dem sich aufbauenden Log die eben eingefügten Einträge.</p>

<h3>Debugging</h3>
<p>Um eine App Schritt für Schritt auszuführen, werden keine weiteren Codezeilen oder Bibliotheken benötigt. Dieses Feature ist direkt in Android Studio eingebaut.  
Einen Haltepunkt können Sie fast überall setzen. Am Meisten Sinn ergibt es bei Anweisungen. Setzen Sie einen Punkt in der Klasse <i>MainApplication</i> in der Zeile, die <code>super.onCreate()</code> enthält. Klicken Sie dafür zwischen der Zeilenzahl und Abtrennung zum Texteditor. Daraufhin wird dort ein roter Punkt eingeblendet. Um das Debugging zu starten, wählen Sie in der Leiste anstelle des <i>Run</i>-Buttons den grünen <i>Käfer</i>, der drei Stellen weiter rechts zu finden ist. Daraufhin erscheint in der Zeile am unteren Rand des Fensters eine neue Option <i>Debug</i>. Wenn Sie das Teilfenster öffnen, sehen Sie im <i>Debugger</i> links den aktuellen Standort und rechts alle Variablen. Steuern können Sie den Vorgang über die Symbole in der oberen Leiste des Debug-Fensters. In der Leiste können Sie die App mit dem <i>Play</i>-Button bis zum nächsten Haltepunkt fortsetzen oder auch mit dem roten Viereck stoppen. Probieren Sie es aus und navigieren Sie in die Tiefen des Codes.</p>

<h2 style="display:none">Lektion 3 - Ein Button</h2>

<p>Bis jetzt ist die App statisch. Alles was verändert wurde, hatte keine sichtbare Auswirkung. Das ändert sich aber in dieser Lektion. Es soll ein Button hinzugefügt werden, der den angezeigten Text verändert.</p>
<p>Öffnen Sie dafür den Layout-Editor, indem Sie die Datei <i>acitivity_main.xml</i> aus <i>app/res/layout</i> anklicken. Es erscheint ein neuer Editor, dessen Benutzeroberfläche im Folgenden erklärt wird:</p>

<img src="images/LayoutGUI.png" />

<p>Ein Layout ist nichts anderes als eine xml-Datei. Beim Wechseln auf die Darstellung <i>Code</i>, kann die zugehörige textuelle Umsetzung betrachtet werden. Das Aussehen jedes Elements wird durch Attribute festgelegt, die entweder im xml-Tag des Elements oder in Design Modus in der rechten Spalte verändert werden können. Es wird grob zwischen zwei verschiedenen Arten von Elementen unterschieden: <i>Layout</i> und <i>Views</i>. Das Layout gibt an, wie die <i>Views</i> angeordnet sind. Diese hingegen können als <i>Widgets</i> verstanden werden, die einer bestimmten Klasse angehören und durch diese unterschiedlich aussehen und funktionieren.</p>

<img src="images/LayoutButton.gif" style="float: right; margin-left: 10px"/>

<p>Wechseln Sie wieder in den Design-Modus, löschen Sie die <i>TextView</i> und ziehen Sie aus der <i>Palette</i> einen <i>Button</i> und eine <i>TextView</i> in eine der beiden Vorschauen. Während die Linke das Aussehen des Layouts in den Vordergrund rückt, ist die rechte Fläche für die Positionierung besonders geeignet. Wird in einer Vorschau etwas geändert, wird die Andere aktualisiert. Richten Sie die beiden Elemente so aus, wie es Ihnen gefällt und drücken Sie dann auf den <i>Zauberstab</i>, der sich über der linken Vorschau befindet. Dadurch werden der Button und der Text von alleine ausgerichtet.</p>

<p>Wählen Sie den Button aus. Im Bereich <i>Attributes</i> sind die Eigenschaften des Elements zu sehen. Ändern Sie die <i>id</i> auf <code>changeText_button</code>. Es ist zu Empfehlen die eindeutigen Ids nach dem Muster <i>aufgabe_art</i> zu benennen. Der Text des Buttons soll <code>Text ändern</code> lauten. Ändern Sie zusätzlich die <i>textSize</i> auf <code>18sp</code>.</p>

<h3>Exkurs: Dimensionen</h3>
<p>Ihnen könnte eben aufgefallen sein, dass die Größe des Textes auf 18<i>sp</i> gesetzt werden sollte. Doch was bedeutet <i>sp</i> überhaupt? Diese und weitere Dimensionseinheiten sollen in diesem Exkurs kurz und bündig eingeführt werden.</p>
<ul>
    <li><i>Density-independent Pixels (dp)</i>: Die Displays der Handys werden immer besser und möchten für ein schöneres Bild immer mehr Pixel anzeigen. Beispielsweise besitzt das Pixel 3 XL mit 523dpi (<i>dots per inch</i>) bei 6,3 Zoll eine geringere Pixeldichte als das Samsung Galaxy S20 mit 563dpi bei 6,2 Zoll, obwohl beide Displays in etwa die gleiche Größe haben. Würden die Dimensionen von Elementen in Pixel angegeben werden, hätte das zur Folge, dass das gleiche Layout auf beiden Geräten unterschiedlich aussehen würde. Deswegen wurde die Einheit <i>dp</i> eingeführt, welche unabhängig von der Dichte der Pixel ist. Bei einer gleichen Displaygröße, aber unterschiedlicher Pixelanzahl, würde die App auf beiden gleich aussehen.</li>
    <li><i>Scale-independent Pixels (sp)</i>: Diese Größeneinheit besitzt alle Eigenschaften von <i>dp</i>. Zusätzlich wird jedoch auch die eingestellte Schriftgröße in die Skalierung eingerechnet. Deswegen eignet sich diese Einheit für Text.</li>
    <li><i>Punkte (pt)</i>: 1pt ist 1/72 Zoll des Bildschirms bei einer Pixeldichte von 72dpi</li>
    <li><i>Pixel (px)</i>: Entspricht den tatsächlich vorhandenen Pixeln auf dem Display.</li>
</ul>

<p>Zusammengefasst lässt sich sagen, dass es viele verschiedene Größeneinheiten gibt, die Anwendung finden könnten. Zusätzlich zu den Vorgestellten sollten noch <i>Millimeter (mm)</i> und <i>Zoll (in)</i> erwähnt werden. Besonders eignen sich aber <i>dp</i> für <i>Layouts</i> und <i>Views</i> und <i>sp</i> für Text, da diese mit der Pixeldichte des Geräts skalieren und dadurch displayübergreifend bei gleicher Größe das gleiche Ergebnis erzielt werden kann.</p>

<p>Der <i>Button</i> ist fertig konfiguriert. Jetzt ist die <i>TextView</i> an der Reihe. Wechseln Sie dafür in die Code-Ansicht. Sie sollten in einem <i>androidx.constraintlayout.widget.ConstraintLayout</i>-Tag einen <i>TextView</i> und einen <i>Button</i>-Tag sehen. Während der Button schon die Attribute zugewiesen bekommen hat, ist die TextView noch relativ leer. Ändern Sie im Code die <i>id</i> zu <code>hello_textView</code>, den <i>Text</i> zu <code>Hello Android</code> und die Größe des Texts auf <code>20sp</code>. Wenn Sie nun in den Design-Modus wechseln, wurden die Änderungen übernommen und die <i>TextView</i> beinhaltet unter Anderem den eingestellten Text.</p>
<p>Jedoch ist sowohl im Code- als auch im Design-Modus der Text gelb hinterlegt. Das liegt daran, dass Android keinen <i>hardcoded text</i> mag. Ist der Text direkt per Hand eingegeben, steigt der Wartungs- und Übersetzungsaufwand. Deswegen gibt es die Möglichkeit den Text durch eine Variable zu ersetzen, der der Text hinterlegt ist. Diese werden in der Datei <i>app/res/values/strings.xml</i> zentral verwaltet. Hier gibt es ebenfalls wieder zwei Möglichkeiten neue Zeichenketten hinzuzufügen:</p>

<img src="images/StringValue.gif" style="float: right; margin-left: 10px" />

<ul>
    <li>Design-Modus (<i>Button</i>): Klicken Sie auf den kleinen Balken neben der Texteingabe im Attributbereich. Es öffnet sich ein Fenster, in dem alle aktuell verfügbaren Zeichenketten angezeigt werden. Aktuell ist für Ihre App nur der Name der App hinterlegt. Fügen Sie eine neue Variable hinzu, indem Sie oben links auf das <i>+</i> drücken und <i>String Value</i> auswählen. Der Name der Variable soll <code>text_aendern</code> lauten. <p>Zuerst brauchen Sie Zugriff auf die beiden Elemente. Dies kann mit Hilfe der eben definierten, eindeutigen Id geschehen. Die Methode <code>findViewById()</code> gibt das Element zurück, dass die übergebe Id besitzt. Zusätzlich muss die Klasse der Rückgabe als generischer Parameter zwischen Methodenname und Parameterklammern angegeben werden. Für den Button ergibt sich dadurch folgender Code:</p>

<blockquote><code>val button = findViewById&#60;Button>(R.id.changeText_button)</code></blockquote>

<p>Mit <i>R</i> können Sie auf die Ressourcen der App zugreifen. Fügen Sie folgenden Code vor der schließenden Klammer von <i>onCreate()</i> ein:</p>

<blockquote><code>button.setOnClickListener {

}</code></blockquote>

<h2 style="display:none">Lektion 4 - View Binding</h2>

<img src="images/View Binding.png" style="float: right; margin-left: 10px; max-width: 450px" />

<p>Das Layout und die Logik sind zwei getrennte Teile der App. Mit <i>findViewById()</i> wurde in der letzten Lektion mit Hilfe einer eindeutigen Bezeichnung auf das Element im Layout von der Logik aus zugegriffen. Im Hintergrund wird dabei während der Ausführung der App die Layouthierarchie, die eine Baumstruktur aufweist, durchlaufen. Dies kann bei vielen Knoten zu einer hohen Laufzeit und einer langsamen App führen. Deswegen hat Android das Konzept des <i>View Bindings</i> eingeführt. Ist es für die App aktiv, wird automatisch für alle Layouts ein <i>Binding Object</i> generiert, das als Schnittstelle fungiert. In diesem sind unter anderem Felder für jedes Layout und Element vorhanden. Das besondere ist, dass das Objekt bei der Kompilierung erzeugt wird und bei Ausführung der App bereits vollständig vorliegt. Bei Ausführung wird das Feld direkt angesprochen und kein Baum durchsucht. Der Name des Objekts leitet sich aus dem Namen der zugehörigen Datei und der Endung <i>Binding</i> ab (beispielsweise für die Datei <i>acitivity_main</i> <i>ActivityMainBinding</i>). Die Schnittstelle wird in <i>onCreate()</i> der Aktivität mit <i>inflate()</i> erzeugt. Im Folgenden wird dann nicht das Layout aus den Ressourcen, sondern das äußerste Layout des Objekts, auf das mit <i>root</i> zugegriffen werden kann, als Inhalt der Aktivität gesetzt.</p>
<p>Ein Vorteil dieses Konzepts ist die Laufzeit. Da das <i>Binding Object</i> bereits bei der Kompilierung generiert wird, muss bei der Ausführung nur noch auf ein Feld des Objekts zugegriffen und nicht mehr der ganze Layoutbaum traversiert werden. Zudem ist es auch sicherer. <i>findViewById()</i> könnte eine <i>NullPointerException</i> werfen, wenn die angegebene Id nicht vergeben ist. Zudem muss auch keine händische Typverengung von <i>View</i> zu beispielsweise <i>Button</i> vorgenommen werden. Die Felder des Objekts haben alle bereits den richtigen Typ. Fügen Sie nun <i>View Binding</i> der bereits bestehenden <i>HelloWorld-App</i> hinzu.</p>

<p>Zu Beginn ist es notwendig das Feature zu aktivieren. Erst dann findet die Generierung der Objekte bei Kompilierung statt. Fügen Sie dazu der Datei <i>build.gradle</i> auf Modulebene in der Sektion <i>android</i> folgenden Block an:</p>

<blockquote><code>buildFeatures {  
   viewBinding true
}</code></blockquote>

<p>Vergessen Sie nicht die Änderungen zu synchronisieren. Jetzt kann auf das Objekt in der <i>MainActivity</i> zugegriffen werden. Fügen Sie der Klasse ein privates Feld, das später initialisiert wird, mit dem Namen <code>binding</code> und dem Typ <code>ActivityMainBinding</code> hinzu.</p>
<blockquote><code>private lateinit var binding: ActivityMainBinding</code></blockquote>

<p>Initialisieren Sie dieses daraufhin in der Methode <i>onCreate()</i> nach dem Aufruf der gleichen Methode der Super-Klasse mit:</p>
<blockquote><code>binding = ActivityMainBinding.inflate(layoutInflater)</code></blockquote>

<p>Als letzten Schritt setzen Sie den Inhalt der Aktivität auf das äußerste Layout des <i>Binding Objects</i>:</p>
<blockquote><code>setContentView(binding.root)</code></blockquote>

<p>Jetzt können Sie die lokalen Variablen für den <i>Button</i> und die <i>TextViews</i> löschen und die Elemente direkt mit <i>binding.idDesElements</i> ansprechen. So wird aus der Zeile <blockquote><code>textView.text = "Hello Kotlin"</code></blockquote> zum Beispiel <blockquote><code>binding.helloTextView.text = "Hello Kotlin"</code></blockquote> Ändern Sie alle Zeilen ab, die Fehler verursachen. Beim Starten der App sollte sich nichts verändert haben. Die Lösung können Sie sich in diesem <a target="_blank" src="https://github.com/Limatationz/kotlin-mooc-App-helloWorld/compare/schritt02-button...schritt03-View Binding">git diff</a> anschauen.</p>

<h2 style="display:none">Lektion 5 - Layouts und Übergänge</h2>

<h4>ConstraintLayout</h4>
<img src="images/Constraint1.gif" style="float: right; margin-left: 10px" />

<p>Zuletzt sollen Sie ein Layout <code>constraint_activity</code> erstellen, dass das <code>ConstraintLayout</code> als Basis besitzt. Das Besondere an diesem ist, dass keine vorgeschriebene Ordnung herrscht, sondern diese durch Beziehungen zwischen Elementen generiert wird. Für dieses Layout ist es vorteilhaft den Design-Modus zu verwenden. Bevor Elemente dem Layout hinzugefügt werden, ist es ratsam einen allgemeinen Abstand zum Rand des Bildschirms einzustellen. Dies können Sie in der Leiste oberhalb der Vorschauen rechts neben dem Magneten einstellen. Aktuell sollte bei Ihnen dort <i>0dp</i> ausgewählt sein. Ändern Sie diesen Wert auf <code>16dp</code>.<br/>
Ziehen Sie einen Button, der die Id <code>weiterConstraint_button</code> besitzen soll, in das Layout. Wenn Sie diesen auswählen, erscheinen vier weiße Punkte an jeder Kante. Wählen Sie den Punkt an der unteren Kante aus und ziehen Sie den Pfeil an den unteren Rand der Vorschau. Wenn Sie die Auswahl nun loslassen, wird der Button an den unteren Rand geschoben. Jedoch nicht komplett. Es wird der vorher eingestellte Abstand von <i>16dp</i> eingehalten. Verbinden Sie jetzt die obere Kante des <i>Buttons</i> mit dem oberen Rand der Vorschau. Das Element wird in die vertikale Mitte geschoben und zwei gezackte Linien erscheinen. Diese drücken aus, dass hier eine relative Beziehung zwischen dem Objekt und dem Ziel des Pfeils besteht. Aktuell besagt diese, dass das Verhältnis vom Abstand nach oben und unten 50/50 beträgt. Das können Sie aber im <i>Attributfenster</i> verändern. Unter dem Punkt <i>Layout</i> wird Ihnen ein Viereck angezeigt. Links von diesem ist ein Schieberegler zu sehen, der auf <i>50</i> steht. Verändern Sie die Stellung des Reglers.<br />
An dem Quadrat sind außerdem mehrere <i>+</i> abgebildet. Diese bedeuten, dass an dieser Kante keine Beziehung existiert. Ist eine zugewiesen, erscheint ein Kasten, indem der Abstand vom Ziel eingestellt werden kann. Wählen Sie einen anderen, höheren Wert aus und schauen Sie was passiert!<br />
Genug der Spielerei. Positionieren Sie den <i>Button</i> in der rechten, unteren Ecke. Er soll zu den Rändern jeweils einen Abstand von <code>16dp</code> besitzen.</p>

<h3>Übergänge</h3>
<p>Die Layouts sind fertig. Im nächsten Schritt wird für jedes Layout eine eigene <i>Activity</i> erstellt, die <i>Buttons</i> mit den <i>TextViews</i> verbunden und Übergänge zwischen den Aktivitäten definiert werden.<br />

In den Methoden soll mit Hilfe von <i>View Binding</i> das entsprechende Layout als Inhalt gesetzt werden. Sie können sich dabei an der Lösung von Lektion 4 orientieren. <blockquote>

    <b>xmlns:tools="http://schemas.android.com/tools"</b>

    android:layout_height="match_parent"

<details>
    <summary>Lösung von <i>HorizontalActivity</i></summary>
<blockquote><code>class HorizontalActivity: AppCompatActivity() {
        private lateinit var binding: HorizontalActivityBinding
        override fun onCreate(savedInstanceState: Bundle?) {
            super.onCreate(savedInstanceState)
            binding = HorizontalActivityBinding.inflate(layoutInflater)
            setContentView(binding.root)
        }
    }</code></blockquote>
</details>

<p>Als letzten Schritt müssen Sie der App noch mitteilen, dass die drei Aktivitäten existieren und diese auch verwendet werden sollen. Dabei muss Eine als Start der App angegeben werden. Dies können Sie in der Datei <i>AndroidManifest.xml</i> erledigen. Aktuell ist die nicht mehr existierende Aktivität <i>MainActivity</i> als Startpunkt angegeben. <blockquote>

   &#60;intent-filter>
      &#60;action android:name="android.intent.action.MAIN" />
      &#60;category android:name="android.intent.category.LAUNCHER" />
   &#60;/intent-filter>
&#60;/activity>
&#60;activity android:name=".VerticalActivity" />
&#60;activity android:name=".ConstraintActivity" /></code>
</blockquote>

<p>Jetzt können Sie sich um die <i>Buttons</i> kümmern. Beginnen Sie mit dem Färben der Flächen. <blockquote>

</blockquote>

<p>Die Farben sollten bei Ihnen rot hinterlegt sein, da sie noch nicht existieren. Fügen Sie die Variablen mit den zugehörigen Hex-Codes in die Datei <i>res/values/colors.xml</i> ein:</p>

<ul>
    <li>red: #F44336</li>
    <li>yellow: #FFEB3B</li>
    <li>blue: #3F51B5</li>
    <li>green: #4CAF50</li>
</ul>

<p>Der <i>Button faerbenHorizontalButton</i> der Klasse <i>HorizontalActivity</i> soll nun einen <i>OnClickListener</i> erhalten. Wenn dieser gedrückt wird, soll die Hintergrundfarbe der <i>farbflaecheHorizontalTextView</i> mit <code>setBackgroundResource(getRandomColor())</code> auf eine zufällige Farbe geändert werden. Fügen Sie den Code ein und probieren Sie es aus.<br />

<blockquote><code>Intent(this, VerticalActivity::class.java)</code></blockquote>

<p>Der Kontext in dem die Operation ausgefürt wird, ist die Aktivität selbst, das Ziel ist die Klasse <i>VerticalActivity</i>. Speichern Sie das Objekt in einer lokalen Variable und übergeben Sie es der Methode <code>startActivity()</code>. Ummanteln Sie die Zeilen mit einem <code>OnClickListener</code> des <i>Weiter-Buttons</i>.</p>

<h2 style="display:none">Lektion 6 - Fragmente</h2>

<p>Bis jetzt wurden für jede neue Ansicht neue Aktivitäten definiert. Jedoch wurde noch nicht näher darauf eingegangen, was eine Aktivität genau ist und welche Aufgaben ihnen zugeordnet werden. Deswegen folgt eine Erklärung in dieser Lektion. Zusätzlich werden sogenannte <i>Fragments</i> eingeführt, die als kleinere Teile einer <i>Activity</i> angesehen werden können:</p>
<ul>
    <li><b>Aktivität</b>: Eine Aktivität beinhaltet die Logik einer Teilaufgabe der App. Zusätzlich ist ihr ein Layout zugeordnet, das definiert wie der Bildschirm aussieht. Mit Hilfe von <i>Layout Inflation</i> werden diese beiden Teile miteinander verbunden. Normalerweise nimmt das Layout den gesamten Bildschirm in Anspruch. Besitzt eine App beispielsweise einen Login, ist es zu empfehlen für diesen eine eigene Aktivität zu definieren und die eigentlichen Inhalte in eine Weitere auszulagern.</li>
    <li><b>Fragment:</b> Fragmente wurden in <i>Android 3.0 Honeycomb</i> eingeführt. Sie sind als austauschbare Inhalte einer Aktivität anzusehen. Dabei kann die Aktivität bestimmte Elemente vorgeben. Denken Sie an <i>Whatsapp</i>. Die untere Leiste, in der die verschiedenen Funktionen zur Verfügung gestellt werden, ist Teil der Aktivität. Der Bereich darüber ist jedoch ein Fragment, das je nach Auswahl mit einem Anderen getauscht wird. Die Aktivität wird dabei aber nicht verlassen. Eine Aktivität kann beliebig viele Fragmente beinhalten. Es ist zu empfehlen für jede Teilaufgabe der Aktivität ein eigenes Fragment zu erstellen. Jedoch kann die ganze App auch nur aus einer Aktivität mit beliebig vielen Fragmenten bestehen. Ein Fragment ist eine eigenständige Einheit und kann von jeglichen Aktivitäten verwendet werden. Fragmente einer Aktivität werden in einem <i>FragmentManager</i> verwaltet. Außerdem kann die Navigation zwischen ihnen mit einem Navigationsgraph ausgestaltet werden.</li>
</ul>

<h3>Exkurs: Back Stack & Fragment Back Stack</h3>
<p>Der <i>Back Stack</i> verwaltet die geöffneten Aktivitäten und stellt die Möglichkeit bereit <i>zurück</i> zu gehen. Beim Öffnen einer neuen Aktivität wird diese auf den Stack gelegt. Wird der <i>Zurück-Button</i> gedrückt, wird die aktuelle Ansicht vom Stack geworfen und die Vorherige wieder geöffnet.<br />
Während der <i>Back Stack</i> Aktivitäten verwaltet, ist der <i>Fragment Back Stack</i> für die Fragmente einer Aktivität zuständig. Wird ein Neues geöffnet, wird dieses auf den Stapel gelegt und bei Aufforderung wieder entfernt.<br />
Beide Stapel haben Auswirkung darauf, wohin zurückgesprungen wird. Gemeinsam haben sie jedoch, dass, wenn das letzte Element vom Stapel genommen wurde, zur letzten App gewechselt wird. Wie sie ineinandergreifen oder wann wohin zurückgesprungen werden soll, kann explizit angegeben werden. Die implizierte Lösung ist, dass zuerst die Fragmente durchgegangen werden und wenn keines mehr vorhanden ist, die Aktivitäten vom Stapel genommen werden. Weitere Informationen finden Sie in der <a target="_blank" href="https://developer.android.com/guide/components/activities/tasks-and-back-stack">Dokumentation</a></p>

<h3>Fragmente</h3>
<p>Fragmente sind eigene, kleinere Einheiten, die in einer Aktivität existieren können. Dabei ist ihnen ein bestimmter Bereich auf dem Bildschirm zugeteilt. Verwaltet werden sie mit einem <i>Fragment Manager</i>. Wandeln Sie im ersten Schritt die bestehenden Aktivitäten in Fragmente um. Ändern Sie die Super-Klasse. Anstelle von <i>AppCompatActivity</i> soll nun von <code>Fragment</code> geerbt werden. Wichtig ist, dass Sie die Klasse aus der Bibliothek <i>androidx</i> importieren. Auf den ersten Blick scheint der Code größtenteils weiterhin zu funktionieren. Jedoch gibt es einige Dinge, die noch angepasst werden müssen.<br />
Da das Fragment im Gegensatz zur Aktivität nicht auf den Kontext, die Inhalte, zum Beispiel die Ressourcen, der App zugreifen kann, muss dieser der <i>onCreateView</i>-Methode übergeben werden. Diese besitzt die gleiche Funktion wie <i>onCreate()</i>. Zusätzlich wird der <i>Layout Inflater</i> benötigt. Zurückgegeben wird die <i>View</i>, also die Ansicht des Fragments, das in den Platzhalter eingefügt werden soll:</p>

<blockquote><code>override fun onCreateView(inflater: LayoutInflater, container: ViewGroup?, savedInstanceState: Bundle?): View?</code></blockquote>

<p>Die zusätzlichen Informationen müssen auch der Methode <i>inflate</i> übergeben werden:</p>
<blockquote><code>binding = HorizontalFragmentBinding.inflate(layoutInflater, container, false)</code></blockquote>

<p>Zum Schluss muss die Hauptansicht zurückgegeben werden:</p>

<blockquote><code>return binding.root</code></blockquote>

<p>Übernehmen Sie die Änderungen angepasst in allen Fragment-Klassen. Kommentieren Sie zudem den Inhalt des <i>OnClickListener</i> der <i>Weiter-Buttons</i> ein.</p>

<p>Jetzt kann zunächst ein Fragment mit der Aktivität verbunden werden. Fügen Sie dem Layout der Aktivität <i>MainActivity</i> ein <i>fragment</i> hinzu. Setzen Sie deren Höhe und Breite auf <code>match_parent</code> und geben Sie ihr die Id <code>fragmentContainerView</code>. Um direkt ein Fragment zuzuweisen, muss dieses im Attribut <i>name</i> angegeben werden: <code>com.example.android.layoutplayground.fragments.HorizontalFragment</code>. Der Code des Layouts sollte jetzt so aussehen:</p>

<h3>Navigation</h3>

<img src="images/NavFolder.gif" style="float: right; margin-left: 10px" />

<p>Aktuell wird nur ein definiertes Fragment angezeigt. Um die Funktionalität aber vollständig zu gewährleisten, fehlt noch der Wechsel zwischen den Fragmenten. Dies wird in diesem Unterabschnitt mit Hilfe eines Navigationsgraph realisiert. Dieser ist, ähnlich zu den Layouts, im Grunde nichts anderes als eine xml-Datei, die aber grafisch angepasst werden kann.<br />
Erstellen Sie für die Datei einen neuen Ordner in <i>res</i>. Wählen Sie dafür die Auswahlmöglichkeit <i>Android Resource Directory</i> im Bereich <i>New</i> aus. Der Name, sowie der Typ der Ressource soll <code>navigation</code> sein. Bevor Sie den Graph erstellen können, müssen Sie die zugehörigen Bibliotheken dem Projekt hinzufügen. Fügen Sie in dem Abschnitt <i>dependencies</i> in <i>build.gradle</i> auf der Modulebene folgende Zeilen ein:</p>
<blockquote><code>
implementation 'androidx.navigation:navigation-fragment-ktx:2.3.5'
implementation 'androidx.navigation:navigation-ui-ktx:2.3.5'
</code></blockquote>

<p>Kopieren Sie zusätzlich die folgende Erweiterung in den Abschnitt <i>plugins</i></p>
<blockquote><code>id 'androidx.navigation.safeargs'</code></blockquote>

<p>Als letztes müssen Sie noch die Quelle des Plugins in die <i>build.gradle</i>-Datei auf der Projektebene im Bereich <i>dependencies</i> angeben:</p>
<blockquote><code>classpath "androidx.navigation:navigation-safe-args-gradle-plugin:2.3.5"</code></blockquote>

<p>Jetzt können Sie die Datei <i>nav_graph.xml</i> im Ordner <i>navigation</i> erstellen. Es öffnet sich ein Editor, der dem von Layouts ähnelt. Im Bereich <i>Host</i> steht eine Fehlermeldung, dass kein <i>NavHostFragment</i> gefunden wurde. Dieses ist in dieser App das <i>fragment</i> des Layouts der Aktivität. Um die beiden Elemente miteinander zu verbinden, müssen weitere Attribute der <i>View</i> hinzugefügt werden:</p>
<blockquote><code>&#60;androidx.constraintlayout.widget.ConstraintLayout xmlns:android="http://schemas.android.com/apk/res/android"
    <b>xmlns:app="http://schemas.android.com/apk/res-auto"</b>
    xmlns:tools="http://schemas.android.com/tools"
    android:layout_width="match_parent"
    android:layout_height="match_parent"
    tools:context=".MainAcitivity">
    
    &#60;fragment
        android:id="@+id/fragmentContainerView"
        <b>android:name="androidx.navigation.fragment.NavHostFragment"</b>
        android:layout_width="match_parent"
        android:layout_height="match_parent"
        <b>app:defaultNavHost="true"</b>
        <b>app:navGraph="@navigation/navigation"</b>
 
&#60;/androidx.constraintlayout.widget.ConstraintLayout></code></blockquote>

<p>Jetzt ist der Fehler im Graph verschwunden und an dessen Stelle ist <i>main_activity</i> zu finden.<br />
Fügen Sie nun die Fragmente der Navigation hinzu. Klicken Sie auf das <i>Handy-Symbol mit dem kleinen grünen Plus in der unteren, rechten Ecke</i>. Dieses steht dafür, dass dies der Startpunkt der Navigation ist und als erstes angezeigt wird. Das Startfragment dieser App soll wieder das horizontale Fragment sein.<br />

<p>Die Übergänge sind definiert, jetzt können sie mit den <i>Buttons</i> verbunden werden. Löschen Sie den Code in <i>setOnClickListener</i> des <i>Weiter-Buttons</i> des <i>HorizontalFragment</i>. Zugriff auf die Navigation kann mit Hilfe des <i>NavControllers</i> erlangt werden. Auf diesem kann die Methode <i>navigate</i> aufgerufen werden, die den Übergang als Argument übergeben bekommt. Dafür wird Dank dem Plugin <i>safe-args</i> ein Hilfsobjekt erstellt, ähnlich wie beim <i>Binding</i> mit der Endung <i>Directions</i>. Dieses enthält alle navigierbaren Pfade des Fragments. Für das <i>HorizontalFragment</i> sieht der passende Code folgendermaßen aus:</p>
<blockquote><code>findNavController().navigate(HorizontalFragmentDirections.actionHorizontalFragmentToVerticalFragment())</code></blockquote>
<p>Passen Sie die restlichen Klassen auf die gleiche Weise an und probieren Sie die App aus.</p>

<h3>Argumente</h3>

<img src="images/AddArgument.gif" style="float: right; margin-left: 10px" />
<p>Den Fragmentwechseln können, wie bei Funktionen, Argumente übergeben werden. Dadurch ist es möglich Informationen von einem zum anderen Fragment zu übertragen. In diesem Beispiel soll das Konzept dafür verwendet werden, dass nach dem Wechsel des Fragments die aktuelle Farbe der Fläche beibehalten werden soll.<br />
Die Argumente können direkt im Navigationsgraphen definiert werden. Dafür steht im Attributfenster der Abschnitt <i>Arguments</i> bereit, wenn ein Element ausgewählt wurde. Soll beispielsweise das <i>HorizontalFragment</i> dem <i>VerticalFragment</i> eine Information übergeben, muss diese dem <i>VerticalFragment</i>, also dem Ziel, hinzugefügt werden. Klicken Sie dafür auf das <i>+</i>. Es öffnet sich ein Fenster, in dem Sie weitere Angaben zu dem Argument bestimmen können. Da die Farbe auf eine Ressource verweist, welche wiederum als Id einen Integer-Wert besitzt, soll das Argument vom Typ <code>Integer</code> sein. Nennen Sie es <code>color</code>. Fügen Sie allen anderen Fragmenten das gleiche Argument hinzu.<br />
Jedem Übergang muss jetzt ein passender Übergabeparameter hinzugefügt werden. Da nicht direkt auf die Hintergrundfarbe der <i>TextView</i> zugegriffen werden kann, muss ein Umweg über eine Variable gegangen werden. Definieren Sie in der Methode <i>onCreate</i> eine veränderbare Variable und initialisieren Sie diese mit <code>-1</code>. Wird der <i>Färben-Button</i> gedrückt, soll zuerst der Variable das Ergebnis der Funktion <i>getRandomColor</i> zugewiesen und daraufhin der Hintergrund der <i>Farbfläche</i> auf den Wert gesetzt werden. Fügen Sie außerdem die Variable als Übergabeparameter beim Fragmentwechsel hinzu.</p>

<details>
    <summary>Lösung für <i>HorizontalFragment</i></summary>
<blockquote><code>override fun onCreateView(inflater: LayoutInflater, container: ViewGroup?,
                              savedInstanceState: Bundle?): View?{
        super.onCreate(savedInstanceState)
        binding = HorizontalFragmentBinding.inflate(layoutInflater, container, false)
 
        var color = -1
 
        binding.faerbenHorizontalButton.setOnClickListener {
            color = getRandomColor()
            binding.farbflaecheHorizontalTextView.setBackgroundResource(color)
        }
 
        binding.weiterHorizontalButton.setOnClickListener {
            findNavController().navigate(HorizontalFragmentDirections.actionHorizontalFragmentToVerticalFragment(color))
        }
 
        return binding.root
    }</code></blockquote>
</details>

<p>Im Zielfragment kann auf die Argumente wieder mit einem generierten Objekt zugegriffen werden, das die Endung <i>Args</i> besitzt. Mit dem Aufruf <code>fromBundle(arguments!!)</code> auf dem Objekt können Sie die Parameter ansprechen:</p>

<blockquote><code>VerticalFragmentArgs.fromBundle(arguments!!).color</code></blockquote>

<p>Speichern Sie den Wert in einer lokalen Variable, testen Sie, ob überhaupt eine Farbe gesetzt wurde (dann ist der Wert ungleich -1) und setzen Sie die Hintergundfarbe der Farbfläche auf diesen.</p>

<details>
    <summary>Lösung für <i>VerticalFragment</i></summary>
<blockquote><code>val startColor = VerticalFragmentArgs.fromBundle(arguments!!).color
if(startColor != -1)
   binding.farbflaecheVerticalTextView.setBackgroundResource(startColor)</code></blockquote>
</details>

<p>Übertragen und passen die den Code für alle anderen Fragmente an. Probieren Sie daraufhin die App aus.<br />
Sie sollten feststellen, dass die App direkt abstürzt. Suchen Sie mit Hilfe des Logs den Fehler und versuchen Sie ihn zu beheben.</p>

<details>
    <summary>Lösung</summary>
    <p>Beim Starten der App wird dem <i>HorizontalFragment</i> kein Argument übergeben. Die Folge ist, dass ein Fehler geworfen wird und die App abstürzt. Eine einfache Lösung ist die involvierten Codezeilen mit einem <code>try and catch-Block</code> zu ummanteln, der den Fehler abfängt.</p>
<blockquote><code>try {
   val startColor = HorizontalFragmentArgs.fromBundle(arguments!!).color
   if (startColor != -1)
      binding.farbflaecheHorizontalTextView.setBackgroundResource(startColor)
}
catch (e: Exception){}</code></blockquote>
</details>

<h2 style="display:none">Lektion 7 - Menüs</h2>
<p>Menüs sind Navigationselemente in einer Aktivität. Einige sind versteckt und werden erst beim Aktivieren einer Schaltfläche angezeigt. Sie bestehen aus verschiedenen <i>Items</i>, die zu bestimmten Fragmenten führen. In Android waren viele Jahre das <i>Options Menu</i> und der <i>Navigation Drawer</i> am verbreitetsten. <h3>Options Menu</h3>
<img src="images/OptionsMenu.gif" style="float: right; margin-left: 10px" />

<h4>Options Menu</h4>

<img src="images/CreateMenu.gif" style="float: right; margin-left: 10px" />

<p>Menüs sind wieder <i>xml-Dateien</i>. Den Ordner, in dem diese Art von Ressourcen gespeichert werden, müssen Sie erst in <i>res</i> hinzufügen. Erstellen Sie dafür ein neues <i>Android Resource Directory</i> mit dem Namen <code>menu</code> und dem Typ <code>menu</code>. Das Menü soll den Dateinamen <code>options_menu</code> besitzen. Wenn Sie die Datei öffnen, erscheint eine bekannte Arbeitsfläche. Ein Unterschied ist, dass weniger und andere Items in der <i>Palette</i> ausgewählt werden können. Fügen Sie dem Menü ein <code>Menu Item</code> hinzu und geben Sie diesem die <i>id</i> <code>backToTitle</code>. Angezeigt werden soll aber der Text <code>Zurück zum Start</code>.</p>
<p>Das Menü ist fertig, jetzt fehlt nur noch die Verbindung mit der Aktivität. Ein <i>Options Menu</i> kann mit der Methode <i>onCreateOptionsMenu</i> erzeugt werden. Da es in jedem Fragment angezeigt werden soll, muss diese Methode in der Aktivität überschrieben werden. Die Implementierung in einem Fragment ähnelt dieser. Zusätzlich müssen Sie aber in <i>onCreate()</i> die Zeile <code>setHasOptionsMenu(true)</code> hinzufügen und der <i>onCreateOptionsMenu</i>-Methode zusätzlich den <i>menuInflator</i> übergeben.<br />
Überschreiben Sie die gerade erwähnte Methode. Verwenden Sie dafür das zugehörige Hilfefenster, das Sie mit <i>Strg+O</i> öffnen können. In der Methode soll das Menü mit einem <i>MenuInflater</i> hinzugefügt werden. Auf diesen können Sie mit dem Feld <code>menuInflater</code> zugreifen. Speichern Sie das Objekt in einer lokalen Variable.<br />
Im nächsten Schritt soll auf dem eben gespeicherten <i>Inflator</i> die bereits bekannte Methode <code>inflate</code> aufgerufen werden. Als Übergabeparameter müssen Sie zuerst die Ressource der eben erstellten Menü-Datei und das <code>menu</code>, das der Methode übergeben wird, angeben. Geben Sie zum Schluss <code>wahr</code> zurück. Ihre Methode sollte jetzt so aussehen:</p>

<blockquote><code>override fun onCreateOptionsMenu(menu: Menu): Boolean {
    val inflater: MenuInflater = menuInflater
    inflater.inflate(R.menu.options_menu, menu)
    return true
}</code></blockquote>

<p>Wenn Sie jetzt die App ausprobieren, wird das Menü in der rechten, oberen Ecke angezeigt und kann auch geöffnet werden. Beim Auswählen des Menüpunkts passiert aber noch nichts. Dafür muss eine zweite Methode überschrieben werden: <i>onOptionsItemSelected</i>. Sie bekommt das ausgewählte <i>MenuItem</i> übergeben und gibt einen Wahrheitswert zurück. Um ein bestimmtes Verhalten bei einem bestimmten <i>item</i> auszulösen, muss zuerst überprüft werden, ob dessen <i>id</i> die gewollte ist. Es ist zu empfehlen <code>when</code> zu verwenden, dem <code>item.id</code> übergeben wird. Auf den eindeutigen Namen des erstellten Menüeintrags kann mit <code>R.id</code> zugeriffen werden. Wenn die Ids übereinstimmen, soll eine neue zu erstellende Methode <code>backToTitle</code> aufgerufen und danach <code>true</code> zurückgegeben werden. Fügen Sie dem <i>when-Block</i> außerdem folgenden Standardfall hinzu: <code>else → super.onOptionsItemSelected(item)</code>. Vollständig implementiert könnte diese Methode wie folgt aussehen:</p>
<blockquote><code>override fun onOptionsItemSelected(item: MenuItem): Boolean {
    return when (item.itemId) {
        R.id.backToTitle → {
            backToTitle()
            true
        }
        else → super.onOptionsItemSelected(item)
    }
}</code></blockquote>

<p>Vervollständigen Sie zum Schluss noch die Methode <i>backToTitle</i>, die Sie mit Hilfe von <i>Context Actions</i> automatisch erstellen lassen können.<br />
Starten Sie die App im Emulator oder auf einem angeschlossenen Gerät und testen Sie, ob das Menü und der Wechsel der Aktivitäten wie erwartet funktioniert. Eine Lösung können Sie in diesem <a target="_blank" href="https://github.com/Limatationz/kotlin-mooc-App-LayoutPlayground/compare/schritt04-navigation...schritt05-optionsMenu">git diff</a> betrachten.</p>


<h3>Navigation Drawer</h3>
<img src="images/NavDrawer.gif" style="float: right; margin-left: 10px" />
<p>Der <i>Navigation Drawer</i> wird meist als Hauptmenü der App verwendet. Erreichen können die Benutzer:innen ihn durch das Symbol in der linken oberen Ecke mit den drei horizontalen Balken, oder indem Sie vom linken Rand nach rechts wischen. Eine Besonderheit ist, dass diesem Menü ein Header angefügt werden kann. In Ihrer App soll diese Navigation die Möglichkeit bieten zwischen den Fragmenten beliebig zu wechseln.<br />
Da der <i>Navigation Drawer</i> fragmentunabhängig agieren soll, muss er in der Aktivität definiert werden. Dazu ist ein neues Hauptlayout nötig, in dem eine <i>NavigationView</i> platziert wird. Ersetzen Sie dazu das <i>ConstraintLayout</i> der <i>main_activity</i> durch ein <code>androidx.drawerlayout.widget.DrawerLayout</code>. Durch den Wechsel des Layouts wird der Aktivität das Grundgerüst des <i>Navigation Drawers</i> mitgegeben. Dieses muss nur noch mit dem Menü gefüllt werden. Das<i>fragment</i> wird dadurch nicht beeinflusst.<br />
Kopieren Sie den folgenden Code in das <i>DrawerLayout</i>:</p>

<blockquote><code>&#60;com.google.android.material.navigation.NavigationView
    android:id="@+id/nav_view"
    android:layout_width="wrap_content"
    android:layout_height="match_parent"
    android:layout_gravity="start"/></code></blockquote>
    
<p>Damit ist das Layout fast fertig.<br /> Es fehlt noch das Menü an sich. Erstellen Sie dazu eine Datei <code>navdrawer_menu.xml</code> im Ordner <i>menu</i>. In dieser soll für jedes Fragment ein <i>Item</i> erstellt werden. Wählen Sie einen passenden Titel, der im Menü angezeigt werden soll. Die <i>id</i> jedes <i>Items</i> muss der <i>id</i> des entsprechenden Fragments im <i>nav_graph</i> entsprechen. Nur dann ist eine Navigation zu diesem möglich.</p>

<details>
    <summary>Mögliche Lösung für <i>navdrawer_menu</i></summary>
<blockquote><code>&#60;?xml version="1.0" encoding="utf-8"?>
&#60;menu xmlns:android="http://schemas.android.com/apk/res/android">
    &#60;item
        android:id="@+id/horizontalFragment"
        android:title="Horizontal" />
    &#60;item
        android:id="@+id/verticalFragment"
        android:title="Vertical" />
    &#60;item
        android:id="@+id/constraintFragment"
        android:title="Constraint" />
&#60;/menu></code></blockquote>
</details>

<p>Im letzten Schritt muss das Menü mit der <i>MainActivity</i> und deren <i>Actionbar</i> verbunden werden. Wandeln Sie zuerst die Klasse soweit um, dass <i>View Binding</i> verwendet wird. Erstellen Sie daraufhin ein privates Feld <code>drawerLayout</code> vom Typ <code>DrawerLayout</code>, das in <i>OnCreate</i> entsprechend initialisiert wird. Zusätzlich soll folgender Code hinzugefügt werden:</p>
<blockquote><code>drawerLayout = binding.root
val navController = this.findNavController(R.id.fragmentContainerView)
NavigationUI.setupActionBarWithNavController(this, navController, drawerLayout)
NavigationUI.setupWithNavController(binding.navView, navController)</code></blockquote>

<p>Wird mit dem Menü zu einem Fragment gesprungen, werden diesem keine Argumente übergeben. Passen Sie den entsprechenden Block in <i>onCreateView</i> der Klassen <i>VerticalFragment</i> und <i>ConstraintFragment</i> dem von <i>HorizontalFragment</i> an, sodass die App nicht abstürzt.<br />
Nach einer Navigation mit Hilfe des <i>Navigation Drawer</i>, verschwindet das <i>Hamburger-Symbol</i> und es erscheint ein <i>Pfeil</i>, der zurück zum Startfragment führen soll. Um diese Funktionalität zu unterstützen, müssen Sie folgende Methode überschreiben:</p>

<blockquote><code>override fun onSupportNavigateUp(): Boolean {
   val navController = this.findNavController(R.id.fragmentContainerView)
   return NavigationUI.navigateUp(navController, drawerLayout)
}</code></blockquote>

<p>Probieren Sie die App aus und verwenden Sie das neu erstelle Menü zur Navigation. Falls Probleme auftreten, können Sie Ihren Code mit einer Lösung mit Hilfe von diesem <a target="_blank" href="https://github.com/Limatationz/kotlin-mooc-App-LayoutPlayground/compare/schritt05-optionsMenu...schritt06-navDrawer">git diff</a> vergleichen.</p>

<h3>Bottom Navigation</h3>
<img src="images/BottomNavigation.gif" style="float: right; margin-left: 10px" />
<p>Anstelle eines <i>NavDrawer</i>s wird in modernen Apps meist eine <i>Bottom Navigation</i> zur Navigation verwendet. Sie eignet sich vor Allem, wenn zwischen 3 und 5 Fragmenten gewechselt werden soll. Ihre Implementierung ähnelt im groben der des <i>NavDrawer</i>s.<br />
    Passen Sie das Layout der <i>MainActivity</i> in soweit an, dass statt dem <i>DrawerLayout</i> ein <i>ConstraintLayout</i> das äußerste Layout darstellt. Zudem soll eine <i>BottomNavigationView</i> hinzugefügt werden:</p>
    
<blockquote><code>&#60;com.google.android.material.bottomnavigation.BottomNavigationView
    android:id="@+id/bottomNavigation"
    android:layout_width="match_parent"
    android:layout_height="wrap_content"
    android:layout_gravity="start"
    app:layout_constraintBottom_toBottomOf="parent"
    app:layout_constraintEnd_toEndOf="parent"
    app:layout_constraintHorizontal_bias="0.5"
    app:layout_constraintStart_toStartOf="parent"
    app:menu="@menu/navdrawer_menu" /></code></blockquote>

<p>Als Menü wird das bereits erstellte <i>navdrawer_menu</i> verwendet.<br />
    Ordnen Sie das <i>fragment</i> über der <i>BottomNavigationView</i> an.<br />
    In der <i>MainActivity</i> können alle Zeilen, die den <i>NavDrawer</i> beziehungsweise das <i>DrawerLayout</i> gelöscht werden. Fügen Sie dafür in <i>onCreate()</i> folgende Zeilen hinzu:</p>
    
<blockquote><code>val navController = this.findNavController(R.id.fragmentContainerView)
binding.bottomNavigation.setupWithNavController(navController)</code></blockquote>

<p>Und schon ist die <i>Bottom Navigation</i> fertig. die Lösung können Sie wie gewohnt in diesem <a href="https://github.com/Limatationz/kotlin-mooc-App-LayoutPlayground/compare/schritt06-navDrawer...schritt07-bottomNav" target="_blank">git diff</a> ansehen.</p>
